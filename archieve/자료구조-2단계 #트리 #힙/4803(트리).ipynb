```python
import os, io, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

def sol(G, T) :
  NOT_FOUND = -1
  vis = [NOT_FOUND] * (N+1)

  def D(G, vis, s) :
    S = [s]
    cyc = False
    p = 0 #방문한 번호
    while S :
      u = S.pop()
      if vis[u] != NOT_FOUND : 
        if vis[u] != p-1 : cyc = True 
        continue

      vis[u] = p
      p += 1
      S += G[u]

    return not cyc

  answer = 0
  for i in range(1, N+1) :
    if vis[i] != NOT_FOUND : continue
    if D(G, vis, i): answer += 1
  
  if not answer :
    sys.stdout.write(f'Case {T}: No trees.\n')
  elif answer == 1 :
    sys.stdout.write(f'Case {T}: There is one tree.\n')
  else :
    sys.stdout.write(f'Case {T}: A forest of {str(answer)} trees.\n')

T = 0
while True :
  T += 1
  N, M = map(int, input().split())
  if M == N == 0 : break

  G = [[] for _ in range(N+1)]
  for _ in range(M) :
    u, v = map(int, input().split())
    G[u].append(v)
    G[v].append(u)
  
  sol(G, str(T))
```

- DFS로 사이클을 찾는 것은, 이미 방문한 노드가 `부모 노드`가 아닌지 확인하는 것이다, 방금 방문한 노드가 아니라.
- `if vis[u] != p-1 : cyc = True` 틀린 로직. 
  - `p-1`는 해당 노드의 부모가 아닐 수 있다.
  - 이전에 방문한 노드와 부모 노드를 같은 것으로 오해했다. 

```python
import os, io, sys
sys.setrecursionlimit(10**6)
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

def sol(G, T) :
  vis = [False] * (N+1)

  def D(G, vis, s, p) : #사이클이 존재하면 True
    vis[s] = True
    for v in G[s] :
      if not vis[v] : 
        if D(G, vis, v, s) : return True
      elif v != p : return True #이미 방문한 노드인데, 그게 부모 노드가 아니면 사이클이 존재한다는 뜻
    return False

  answer = 0
  for i in range(1, N+1) :
    if vis[i] : continue
    if not D(G, vis, i, -1): answer += 1
  
  if not answer :
    sys.stdout.write(f'Case {T}: No trees.\n')
  elif answer == 1 :
    sys.stdout.write(f'Case {T}: There is one tree.\n')
  else :
    sys.stdout.write(f'Case {T}: A forest of {str(answer)} trees.\n')

T = 0
while True :
  T += 1
  N, M = map(int, input().split())
  if M == N == 0 : break

  G = [[] for _ in range(N+1)]
  for _ in range(M) :
    u, v = map(int, input().split())
    G[u].append(v)
    G[v].append(u)
  
  sol(G, str(T))
```

### 정해
- 부모 노드를 매 재귀마다 확인해서 부모 노드를 확실하게 파악한다.

```python
import os, io, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

def sol(G, T) :
  vis = [False] * (N+1)

  def D(G, s) :
    S = [s]
    P = [-1]
    vis[s] = True
    cyc = False
    while S :
      u, p = S.pop(), P.pop()
      for v in G[u] :
        if not vis[v]:
          vis[v] = True
          S.append(v) #next node
          P.append(u) #previous parent
        elif v != p : cyc = True
    return cyc

  answer = 0
  for i in range(1, N+1) :
    if vis[i] : continue
    if not D(G, i): answer += 1
  
  if not answer :
    sys.stdout.write(f'Case {T}: No trees.\n')
  elif answer == 1 :
    sys.stdout.write(f'Case {T}: There is one tree.\n')
  else :
    sys.stdout.write(f'Case {T}: A forest of {str(answer)} trees.\n')

T = 0
while True :
  T += 1
  N, M = map(int, input().split())
  if M == N == 0 : break

  G = [[] for _ in range(N+1)]
  for _ in range(M) :
    u, v = map(int, input().split())
    G[u].append(v)
    G[v].append(u)
  
  sol(G, str(T))
```

### 재귀를 없앤 버전

```python
def D(G, s):
    S = [(s, -1)]
    vis[s] = True
    cyc = False
    while S:
      u, p = S.pop()
      for v in G[u]:
        if not vis[v]:
          vis[v] = True
          S.append((v, u))
        elif v != p:
          cyc = True
    return cyc
```
