```python
import io, os
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def bipartite(G, N, M) :
  def dfs(x) :
    if vis[x]: return False

    vis[x] = True
    for y in G[x]:
      if sel[y] == -1 :
        sel[y] = x
        return True
      
    for y in G[x] :
      if dfs(sel[y]):
        sel[y] = x
        return True
    return False

  sel = [-1] * M
  res = 0
  for i in range(N) :
    vis = [False] * N
    if dfs(i) : res += 1
  return res, sel

def sol():
  N = int(input())
  V = N * 2
  G = [[] for _ in range(V)]

  sharks = []
  for i in range(N) :
    sharks.append([i, tuple(map(int, input().split()))])
  
  for i, x in sharks :
    for j, y in sharks :
      if i == j : continue
      if x == y :
        if i > j : # 능력치가 모두 같을경우 우선순위를 정해주어 모두 공멸하는 경우를 방지한다.
          G[i].append(j)
          G[i + N].append(j)
        else :
          G[j].append(i)
          G[j + N].append(i)
      elif all(x[k] >= y[k] for k in range(3)) :
        G[i].append(j)
        G[i + N].append(j)
        
  ans, sel = bipartite(G, V, V)
  print(N-ans)

sol()
```

### 틀렸던 이유
- 모든 능력치가 같은 상어가 있을 수 있다. 
  - 지문에 이미 잡아먹힌 상어는 다른 상어를 잡아먹을 수 없고, 잡아먹히는 동시에 나머지 상어를 잡아먹을 수 없으므로 서로 먹혀 사라지는 경우는 없어야 한다.
  - 따라서 이 경우 우선순위를 정해 한명이 살아남을 수 있도록 정해줘야 한다.
