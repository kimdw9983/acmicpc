```python
import sys, collections
input=sys.stdin.readline

def sol() :
  N, M, K = map(int, input().split())
  G = [[*input().rstrip()] for _ in range(N)]

  res = []
  for i in range(0, N, K) :
    for j in range(0, M, K) :
      slice = [row[j:j+K] for row in G[i:i+K]]
      res.append(slice)
  
  ans = [[] for _ in range(K)]
  ans2 = 0
  for i in range(K) :
    for j in range(K) :
      x = collections.Counter([res[k][i][j] for k in range(len(res))])
      ans[i].append(x.most_common(1)[0][0])
      ans2 += len(res) - x.most_common(1)[0][1]

  sys.stdout.write(str(ans2) + '\n')
  G2 = [['' for _ in range(M)] for _ in range(N)]
  for i in range(N) :
    for j in range(M) :
      G2[i][j] = ans[i%K][j%K]
  
  for l in G2 :
    sys.stdout.write(''.join(l) + '\n')

sol()
```

### 풀이
1. K * K로 타일을 나눈다.
  - 
2. 각 타일의 n번째 칸들을 비교해서 알파벳 개수를 비교한다.
  - 
3. 가장 많이 등장한 알파벳으로 교체할 것이다. 그렇지 않은 알파벳들의 합을 구해두고, 가장 많이 등장한 알파벳도 저장해둔다.
4. 기존 입력 크기에 맞게 알파벳들을 출력한다.
