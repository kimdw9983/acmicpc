### 아이디어
- (관찰) 2번 쿼리 `2 a b`에서 `L[a] == L[b]` 라면 그 사이의 숫자가 모두 같아진다.
- 모든 원소를 동일하게 만드려면 적어도 3번의 쿼리가 필요하다.
  - A는 순열이므로, 가장 넓은 범위인 `2 1 N`을 수행해도 수열 양 끝의 원소는 반드시 다르다.(그리디) \
  이 숫자를 바꾸기 위해선 1번 쿼리와 `2 1 N`을 한번 더 수행해야만 한다.
- 위 사실에 의해, 3번째나 4번째 쿼리가 반드시 `2 1 N`로 끝나야 한다.
  - 또한, `2 1 N`이 모든 같은 원소로 바꾸게 하는 방법은 $L[1] = L[N]$ 인 방법밖에 없다.
- 그렇다면, 잘 생각해보면 수행할 수 있는 쿼리의 경우의 수가 많지 않다는 것을 알 수 있다. 이 경우를 모두 수행해봐도 제한시간안에 풀 수 있을 것이다.

```python
import io, os
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def q1(L, i) :
    L[i+1], L[i] = L[i], L[i+1]
  
def q2(L, s, e) : #[s, e)
  sv, ev = L[s], L[e-1]
  for k in range(s+1, e) :
    if max(sv, ev) < L[k] :
      L[k] = min(sv, ev)
    if min(sv, ev) > L[k] :
      L[k] = max(sv, ev)
  
def simul(L, ans) :  
  if L[0] == L[-1] :
    return True

  if L[1] == L[-1] :
    ans.append((1, 1))
    return True

  if L[0] == L[-2] :
    ans.append((1, len(L)-1))
    return True
  
  if L[1] == L[-2] :
    ans.append((1, 1))
    ans.append((1, len(L)-1))
    return True

def check(L, ans) :
  if simul(L, ans) and len(ans) < 4 :
    ans.append((2, 1, len(L)))
    print_ans(ans)
    return True

def print_ans(ans) :
  print(len(ans))
  for a in ans :
    print(*a)

def sol() :
  N = int(input())
  L = list(map(int, input().split()))

  T = L[:]
  q2(T, 0, N)
  if check(T, [(2, 1, N)]): return

  T = L[:]
  q1(T, 0)
  q2(T, 0, N)
  if check(T, [(1, 1), (2, 1, N)]): return

  T = L[:]
  q1(T, N-2)
  q2(T, 0, N)
  if check(T, [(1, N-1), (2, 1, N)]): return

  T = L[:]
  q2(T, 1, N)
  if check(T, [(2, 2, N)]): return

  T = L[:]
  q2(T, 0, N-1)
  if check(T, [(2, 1, N-1)]): return

  print(-1)

sol()
```
