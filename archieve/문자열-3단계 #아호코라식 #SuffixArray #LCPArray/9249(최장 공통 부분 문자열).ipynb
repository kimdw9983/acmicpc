```python
import io, os
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def SAIS(S1):
  n = len(S1)
  buckets = [0] * (max(S1) + 2)
  for a in S1:
    buckets[a + 1] += 1
  for b in range(1, len(buckets)):
    buckets[b] += buckets[b - 1]
  isL = [1] * n
  for i in reversed(range(n - 1)):
    isL[i] = +(S1[i] > S1[i + 1]) if S1[i] != S1[i + 1] else isL[i + 1]

  def induced_sort(LMS):
    SA = [-1] * (n)
    SA.append(n)
    endpoint = buckets[1:]
    for j in reversed(LMS):
      endpoint[S1[j]] -= 1
      SA[endpoint[S1[j]]] = j
    startpoint = buckets[:-1]
    for i in range(-1, n):
      j = SA[i] - 1
      if j >= 0 and isL[j]:
        SA[startpoint[S1[j]]] = j
        startpoint[S1[j]] += 1
    SA.pop()
    endpoint = buckets[1:]
    for i in reversed(range(n)):
      j = SA[i] - 1
      if j >= 0 and not isL[j]:
        endpoint[S1[j]] -= 1
        SA[endpoint[S1[j]]] = j
    return SA

  isLMS = [+(i and isL[i - 1] and not isL[i]) for i in range(n)]
  isLMS.append(1)
  LMS = [i for i in range(n) if isLMS[i]]
  if len(LMS) > 1:
    SA = induced_sort(LMS)
    LMS2 = [i for i in SA if isLMS[i]]
    prev = -1
    j = 0
    for i in LMS2:
      i1 = prev
      i2 = i
      while prev >= 0 and S1[i1] == S1[i2]:
        i1 += 1
        i2 += 1
        if isLMS[i1] or isLMS[i2]:
          j -= isLMS[i1] and isLMS[i2]
          break
      j += 1
      prev = i
      SA[i] = j
    LMS = [LMS[i] for i in SAIS([SA[i] for i in LMS])]
  return induced_sort(LMS)

def KASAI(S1, SA):
  n = len(S1)
  rank = [0] * n
  for i in range(n):
    rank[SA[i]] = i
  LCP = [0] * (n - 1)
  k = 0
  for i in range(n):
    SAind = rank[i]
    if SAind == n - 1:
      continue
    j = SA[SAind + 1]
    while i + k < n and S1[i + k] == S1[j + k]:
      k += 1
    LCP[SAind] = k
    k -= k > 0
  return LCP

def sol() :
  S1 = input().rstrip()
  S2 = input().rstrip()

  concat = S1 + b"1" + S2
  SA = SAIS(concat)
  LCP = KASAI(concat, SA)
  m =(0,0)
  for i, j in enumerate(LCP):
    if 0<=SA[i-1] + j - 1<len(S1) and len(S1) < SA[i] + j-1 <len(concat): #
      m = max(m,(j,i))
    if 0<=SA[i] + j - 1<len(S1) and len(S1) < SA[i-1] + j-1 <len(concat):
      m = max(m,(j,i))
  res, idx = m
  print(res)
  print(concat[SA[idx]:SA[idx]+res].decode(encoding="ascii"))

sol()
```
