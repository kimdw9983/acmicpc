```python
import io, os
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def sol() :
  N = int(input())
  L = list(map(int, input().split()))
  rcmp = sorted(set(L))
  cmp = {v: i for i, v in enumerate(rcmp)}
  L = [cmp[v] for v in L]
  
  M = int(input())
  Q = [(s-1, e, i) for i, (s, e) in enumerate([map(int, input().split()) for _ in range(M)])]
  Q.sort(key=lambda x: (x[0] // 2000, x[1]))
  
  lenc = len(rcmp)
  res = [0] * M
  cnt = [0] * lenc #각 숫자의 등장횟수
  ans = 0 

  s = e = 0
  for ns, ne, idx in Q:
    while ns < s:
      s -= 1
      ans += cnt[L[s]] == 0
      cnt[L[s]] += 1
    while e < ne: 
      ans += cnt[L[e]] == 0
      cnt[L[e]] += 1
      e += 1
    while ns > s: 
      cnt[L[s]] -= 1
      ans -= cnt[L[s]] == 0
      s += 1
    while e > ne: 
      e -= 1
      cnt[L[e]] -= 1
      ans -= cnt[L[e]] == 0
    
    res[idx] = ans
  
  print('\n'.join(map(str, res)))

sol()
```
