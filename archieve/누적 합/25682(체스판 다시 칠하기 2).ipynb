```python
import sys
input = sys.stdin.readline

N, M, K = map(int, input().split())

acc_w = [[0] * (M + 1) for _ in range(N + 1)]
acc_b = [[0] * (M + 1) for _ in range(N + 1)]
for i in range(N) :
  w = b = 0
  for j, c in enumerate([*input().rstrip()]):
    if j % 2 :
      b += 1 if c == "B" else 0
      w += 1 if c == "W" else 0
    else :
      b += 1 if c == "W" else 0
      w += 1 if c == "B" else 0
    acc_w[i+1][j+1] = acc_b[i][j+1] + w
    acc_b[i+1][j+1] = acc_w[i][j+1] + b

result = 0
for i in range(N - K + 1) :
  for j in range(M - K + 1) :
    sum1 = (acc_w[i+K][j+K] - (acc_b[i][j+K] if K % 2 else acc_w[i][j+K])) - (acc_w[i+K][j] - (acc_b[i][j] if K % 2 else acc_w[i][j]))
    sum2 = (acc_b[i+K][j+K] - (acc_w[i][j+K] if K % 2 else acc_b[i][j+K])) - (acc_b[i+K][j] - (acc_w[i][j] if K % 2 else acc_b[i][j]))
    result = max(result, sum1, sum2)

answer = K ** 2 - result
print(answer)
```

- 2차원 누적합의 성질을 스스로 알아내는게 너무 오래걸리고 고통스러운 일이였다.

```python
import sys
input = sys.stdin.readline

N, M, K = map(int, input().split())

chess = []
for _ in range(N) :
  chess.append([*input().rstrip()])

acc = [[0] * (M - K + 1) for _ in range(N)]
for i, l in enumerate(chess) :
  for j in range(M - K + 1) :
    w = b = 0
    for k, v in enumerate(l[j:j+K], j) :
      if k % 2 :
        b += 1 if v == "B" else 0
        w += 1 if v == "W" else 0
      else :
        b += 1 if v == "W" else 0
        w += 1 if v == "B" else 0

    acc[i][j] = (w, b)

result = 0
for i in range(N - K + 1) :
  for j in range(M - K + 1) :
    sum1 = sum([acc[x][j][x%2] for x in range(i, i+K)])
    sum2 = sum([acc[x][j][int(not x%2)] for x in range(i, i+K)])
    result = max(result, sum1, sum2)

answer = K ** 2 - result
print(answer)
```
