```python
import io, os, sys
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

INF = 1234567891
def fw(V, E):
  D = [[0 if a == b else INF for a in range(V)] for b in range(V)]
  par = [[None] * V for _ in range(V)]

  for u, v, w in E:
    D[u][v] = w

  for m in range(V):
    for a, l in enumerate(D):
      for b, v in enumerate(l):
        if l[m] + D[m][b] >= v: continue
        l[b] = l[m] + D[m][b]
        par[a][b] = par[m][b]

  return D

import collections
INF = 1 << 30
class MCMF:
  def __init__(self, n) :
    self.len = n
    self.G = [[] for _ in range(n)]
    self.C = [[0] * n for _ in range(n)]
    self.F = [[0] * n for _ in range(n)]
    self.W = [[0] * n for _ in range(n)]

  def addEdge(self, u, v, c, w) :
    self.G[u].append(v)
    self.G[v].append(u)
    self.C[u][v] = c
    self.W[u][v] = w
    self.W[v][u] = -w

  def SPFA(self, src, snk):
    mc = mf = 0
    while True :
      vis = [False] * self.len
      dist = [INF] * self.len
      par = [-1] * self.len

      Q = collections.deque([src])
      dist[src] = 0
      vis[src] = True
      while Q :
        u = Q.popleft()
        vis[u] = False
        for v in self.G[u] :
          if self.C[u][v] - self.F[u][v] > 0 and dist[v] > dist[u] + self.W[u][v] :
            dist[v] = dist[u] + self.W[u][v]
            par[v] = u
            if not vis[v] :
              vis[v] = True
              Q.append(v)

      if par[snk] == -1 : break

      flow = INF
      u = snk
      while u != src :
        flow = min(flow, self.C[par[u]][u] - self.F[par[u]][u])
        u = par[u]
      
      u = snk
      while u != src :
        mc += flow * self.W[par[u]][u]
        self.F[par[u]][u] += flow
        self.F[u][par[u]] -= flow
        u = par[u]

      mf += 1

    return mc, mf

def sol() :
  N, M = map(int, input().split())
  E = []
  for _ in range(M) :
    u, v = map(int, input().split())
    E.append((u-1, v-1, 1))
    E.append((v-1, u-1, 1))
  Vcolor = [*map(int, input().split())]
  coin = [*map(int, input().split())]

  D = fw(N, E) #각 정점으로 가는데 드는 비용
  V = N + M + 2
  src, snk = V-2, V-1
  mcmf = MCMF(V)

  for i, a in enumerate(coin) : #플로우 = 동전간의 스왑으로 생각할 수 있다.
    mcmf.addEdge(src, i, INF, 0)
    for j, b in enumerate(Vcolor[i+1:], i+1) : 
      if a == Vcolor[i] : continue
      if a == b :
        print(i, j, D[i][j])
        mcmf.addEdge(i, N+j, 1, D[i][j]) #최소비용으로 스왑할 수 있는 간선 추가
  
  for i, a in enumerate(Vcolor) :
    mcmf.addEdge(N+i, snk, INF, 0)
  
  mc, mf = mcmf.SPFA(src, snk)
  print(mc, mf)

for _ in range(int(input())) :
  sol()
  print("")
```

### 푸는중
- 각 정점사이의 거리 = w (fw로 각 정점끼리 거리 구함)
- 매칭 = 동전 스왑

```python
import sys, io, os
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

INF = 1 << 30
def fw(V, D):
  par = [[None] * V for _ in range(V)]

  for m in range(V):
    for a, l in enumerate(D):
      for b, v in enumerate(l):
        if l[m] + D[m][b] >= v: continue
        l[b] = l[m] + D[m][b]
        par[a][b] = par[m][b]

  return D

import collections
class MCMF:
  def __init__(self, n) :
    self.len = n
    self.G = [[] for _ in range(n)]
    self.C = [[0] * n for _ in range(n)]
    self.F = [[0] * n for _ in range(n)]
    self.W = [[0] * n for _ in range(n)]

  def addEdge(self, u, v, c, w) :
    self.G[u].append(v)
    self.G[v].append(u)
    self.C[u][v] = c
    self.W[u][v] = w
    self.W[v][u] = -w

  def SPFA(self, src, snk):
    mc = mf = 0
    while True :
      vis = [False] * self.len
      dist = [INF] * self.len
      par = [-1] * self.len

      Q = collections.deque([src])
      dist[src] = 0
      vis[src] = True
      while Q :
        u = Q.popleft()
        vis[u] = False
        for v in self.G[u] :
          if self.C[u][v] - self.F[u][v] > 0 and dist[v] > dist[u] + self.W[u][v] :
            dist[v] = dist[u] + self.W[u][v]
            par[v] = u
            if not vis[v] :
              vis[v] = True
              Q.append(v)

      if par[snk] == -1 : break

      flow = INF
      u = snk
      while u != src :
        flow = min(flow, self.C[par[u]][u] - self.F[par[u]][u])
        u = par[u]
      
      u = snk
      while u != src :
        mc += flow * self.W[par[u]][u]
        self.F[par[u]][u] += flow
        self.F[u][par[u]] -= flow
        u = par[u]

      mf += 1

    return mc, mf

def sol() :
  N, M = map(int, input().split())
  D = [[0 if a == b else INF for a in range(N)] for b in range(N)]
  for _ in range(M) :
    u, v = map(int, input().split())
    D[u-1][v-1] = 1
    D[v-1][u-1] = 1
  Vcolor = [*map(int, input().split())]
  coin = [*map(int, input().split())]

  D = fw(N, D) #각 정점으로 가는데 드는 비용
  V = N * 2 + 2
  src, snk = V-2, V-1
  mcmf = MCMF(V)

  for i, a in enumerate(coin) : #플로우 = 동전간의 스왑으로 생각할 수 있다.
    if a : continue
    mcmf.addEdge(src, i, 1, 0)
    for j, b in enumerate(Vcolor) : 
      if b : continue
      mcmf.addEdge(i, N+j, 1, D[i][j]) #최소비용으로 스왑할 수 있는 간선 추가
      mcmf.addEdge(N+j, snk, 1, 0)
  
  mc, mf = mcmf.SPFA(src, snk)
  return str(abs(mc))

for _ in range(int(input())) :
  sys.stdout.write(sol() + '\n')
```

### 풀이
- 접근은 거의 맞았다. 매칭 비용은 간선 사이의 거리이며, 매칭이 의미하는 것은 동전간의 스왑이다.
- 이분 그래프를 구성하는 과정에서 오류가 있었다. 동전색과 정점색이 같은 것을 매칭해야 했는데 동전색과 정점색이 다른 것을 매칭했다.
  - 어디서 출발하고 어디로 도착하는지에 대한 정보를 좀더 정확히 생각해야겠다. 반대로 썼을줄이야.
- `V`를 N+M+2로 해서 한참동안 MLE를 맞았다.
