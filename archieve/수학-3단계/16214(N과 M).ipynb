- a와 m이 서로소라면 $a^n \equiv a^{n \mod \phi(m)} \mod m$ 이다.
  - 따라서 문제의 정답을 $f(n, m)$ 이라 한다면, n과 m이 서로소일 때 $f(n,m) \equiv n^{f(n, \phi(m))} \mod m$ 이다.

```python
import io, os, sys
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def phi(N) :
  res = N

  i = 2
  while i * i <= N:
    if N % i == 0 :
      res -= res // i
      while N % i == 0: N //= i
    i += 1
  
  if N > 1 : res -= res // N
  return res

def pow(X, Y, M) :
  res = 1
  while Y :
    if Y & 1 : res = res * X % M
    X = X * X % M
    Y >>= 1
  return res

def f(N, M) :
  if N == 1 or M <= 2: return N % M

  size = 1
  PHI = [M] + [0] * 49
  for i in range(1, 50) :
    PHI[i] = phi(PHI[i-1])
    if PHI[i] == 1 : 
      size = i
      break

  L = [0] * 50
  L[size] = 1
  for i in range(size-1, -1, -1) :
    L[i] = pow(N, L[i+1], PHI[i]) + PHI[i]
  L[0] = pow(N, L[1], M)

  return L[0]

def sol() :
  N, M = map(int, input().split())
  return f(N, M)

ans = []
for _ in range(int(input())) :
  ans.append(sol())

sys.stdout.write("\n".join(map(str, ans)))
```
