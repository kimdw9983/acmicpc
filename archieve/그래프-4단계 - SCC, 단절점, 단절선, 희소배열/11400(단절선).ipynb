```python
import io, os, sys
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def AE(G):
  V = len(G)
  D = [0] * V #discovery time
  low = [0] * V #lowest reachable vertex
  parent = [-1] * V 
  AE = set()

  cnt = 1
  s = [*range(V)]
  while s:
    v = s.pop()
    if v < 0:  # DFS subtree의 끝일 경우
      v = ~v  #원본 정점 번호
      u = parent[v]
      if low[v] > D[u]: # u가 v의 부모이고, v의 자식 노드중 어느 것도 부모 노드보다 위로 올라갈 수 없으면 그 간선은 단절선이다
        AE.add((min(u, v), max(u, v)))
      low[u] = min(low[u], low[v])

    if D[v]: continue

    D[v] = cnt
    low[v] = cnt
    cnt += 1

    s.append(~v)
    for w in G[v]:
      if w == parent[v]: continue
      if not D[w]:
        parent[w] = v
        s.append(w)
      else :
        low[v] = min(low[v], D[w])

  return AE

def sol() :
  V, E = map(int, input().split())
  G = [[] for _ in range(V)]

  for _ in range(E) :
    u, v = map(int, input().split())
    G[u-1].append(v-1)
    G[v-1].append(u-1)
  
  res = AE(G)
  ans = []
  for s, e in sorted(res) :
    ans.append(f"{s+1} {e+1}")
  sys.stdout.write(f"{len(res)}\n")
  sys.stdout.write('\n'.join(ans))
sol()
```
