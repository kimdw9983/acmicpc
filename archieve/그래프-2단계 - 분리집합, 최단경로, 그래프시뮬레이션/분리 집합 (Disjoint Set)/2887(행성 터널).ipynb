### 관찰
- N이 10만. Naive하게 풀 순 없다.
- 연결 비용이 min(각 좌표의 차이)로 주어진다. 무언가 greedy하게 접근할 수 있을 것 같다.

### 아이디어
- prim의 아이디어를 빌려, 가장 거리가 짧은 정점끼리 greedy하게 잇다가 이미 연결한 정점이면 제외시키는 방법으로 구하면 되지 않을까?
- 기준행성 하나를 잡고 그것과 상대적인 거리가 짧은 기준으로 정렬하면 될 것 같다는 생각.
  - 아무 기준점이나 잡아도 optimal한 해가 나올까? 그리고 그렇다면 어떻게 증명할 수 있을까?

```python
import io, os, heapq
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

class UF:
  def __init__(self, n):
    self.p = list(range(n))

  def find(self, a):
    a_ = a
    while a != self.p[a]:
      a = self.p[a]
    while a_ != a:
      self.p[a_], a_ = a, self.p[a_]
    return a

  def merge(self, a, b):
    self.p[self.find(b)] = self.find(a)

def sol() :
  N = int(input())
  L = [tuple(map(int, input().split())) for _ in range(N)]
  uf = UF(N)

  x, y, z = L.pop() #기준이 되는 정점
  print(x, y, z)
  A, B, C = [], [], []
  for i, (a, b, c) in enumerate(L, 1):
    heapq.heappush(A, (abs(x - a), i))
    heapq.heappush(B, (abs(y - b), i))
    heapq.heappush(C, (abs(z - c), i))

  ans = 0
  while A and B and C: 
    #3개의 heap중에서 중에서 가장 작은것이 어떤건지 확인한 다음 pop
    a, b, c = A[0], B[0], C[0]
    if a[0] <= b[0] and a[0] <= c[0]:
      x, i = heapq.heappop(A)
      if uf.find(i) == uf.find(0): continue #이미 연결되어있는 경우
      uf.merge(0, i)
      ans += x
    elif b[0] <= a[0] and b[0] <= c[0]:
      y, i = heapq.heappop(B)
      if uf.find(i) == uf.find(0): continue
      uf.merge(0, i)
      ans += y
    else:
      z, i = heapq.heappop(C)
      if uf.find(i) == uf.find(0): continue
      uf.merge(0, i)
      ans += z
  
  print(uf.p)
  print(ans)
  
sol()
```

### 틀린 접근
- 아무래도 잡는 점에 따라서 최적해가 달라지는 것 같다. 그림을 그려보니 오히려 직관적으로 그렇다는걸 알 수 있었다.
- 이쯤에서 분류 힌트를 보니, 정렬, 스패닝트리가 나왔다. 스패닝은 생각했지만 정렬 기준을 좀더 생각해봐야겠다.

```python
import io, os, heapq
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

class UF:
  def __init__(self, n):
    self.p = list(range(n))

  def find(self, a):
    a_ = a
    while a != self.p[a]:
      a = self.p[a]
    while a_ != a:
      self.p[a_], a_ = a, self.p[a_]
    return a

  def merge(self, a, b):
    self.p[self.find(b)] = self.find(a)

def sol() :
  N = int(input())
  L = [tuple(map(int, input().split())) for _ in range(N)]
  uf = UF(N)

  A, B, C = [], [], []
  for i, (a, b, c) in enumerate(L) :
    A.append((a, i))
    B.append((b, i))
    C.append((c, i))
  
  A.sort()
  B.sort()
  C.sort()

  X, Y, Z = [], [], []
  for i in range(N-1) :
    heapq.heappush(X, (abs(A[i][0] - A[i+1][0]), A[i][1], A[i+1][1]))
    heapq.heappush(Y, (abs(B[i][0] - B[i+1][0]), B[i][1], B[i+1][1]))
    heapq.heappush(Z, (abs(C[i][0] - C[i+1][0]), C[i][1], C[i+1][1]))

  ans = 0
  while X and Y and Z :
    dx, x1, x2 = X[0]
    dy, y1, y2 = Y[0]
    dz, z1, z2 = Z[0]

    if dx <= dy and dx <= dz :
      heapq.heappop(X)
      if uf.find(x1) != uf.find(x2) :
        uf.merge(x1, x2)
        ans += dx
    elif dy <= dx and dy <= dz :
      heapq.heappop(Y)
      if uf.find(y1) != uf.find(y2) :
        uf.merge(y1, y2)
        ans += dy
    else :
      heapq.heappop(Z)
      if uf.find(z1) != uf.find(z2) :
        uf.merge(z1, z2)
        ans += dz
  
  print(ans)

sol()
```

### 풀이
- 2번째 풀이를 틀리고 마지막에 어렴풋이 든, 정렬한 것을 또 정렬한다는 아이디어가 맞았던가보다.
  - 각각을 정렬한 뒤, 각 값의 차이를 기준으로 또 정렬한다. 이러면 매번 가장 x, y, z의 각각의 차이값이 작은 순서대로 뺄 수 있다. 
  - 좀 느린걸 보니 더 효율적인 구현이 있는듯하다.

```python
import io, os
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

class UF:
  def __init__(self, n):
    self.p = list(range(n))

  def find(self, a):
    a_ = a
    while a != self.p[a]:
      a = self.p[a]
    while a_ != a:
      self.p[a_], a_ = a, self.p[a_]
    return a

  def merge(self, a, b):
    self.p[self.find(b)] = self.find(a)

def kruskal(n, U, V, W): 
  uni = UF(n)
  c, merge_cnt = 0, 0
  mst_u, mst_v = [], []
  L = sorted(range(len(W)), key=lambda x: W[x]) 
  for i in range(len(W)):
    u, v = U[L[i]], V[L[i]]
    fu, fv = uni.find(u), uni.find(v)
    if fu == fv: continue

    c += W[L[i]]
    merge_cnt += 1
    uni.p[fv] = fu
    mst_u.append(u), mst_v.append(v) 

  return c

def sol() :
  N = int(input())
  L = [tuple(map(int, input().split())) for _ in range(N)]

  A, B, C = [], [], []
  for i, (a, b, c) in enumerate(L) :
    A.append((a, i))
    B.append((b, i))
    C.append((c, i))
  
  A.sort()
  B.sort()
  C.sort()

  U, V, W = [], [], []
  for i in range(N-1) :
    U.append(A[i][1])
    V.append(A[i+1][1])
    W.append(abs(A[i][0] - A[i+1][0]))
    U.append(B[i][1])
    V.append(B[i+1][1])
    W.append(abs(B[i][0] - B[i+1][0]))
    U.append(C[i][1])
    V.append(C[i+1][1])
    W.append(abs(C[i][0] - C[i+1][0]))

  print(kruskal(N, U, V, W))

sol()
```

- 조금더 빠른 구현. A, B, C의 정점 연결의 후보들을 모두 구한뒤 kruskal에 넣는다.
- 내 구현도 heapq가 들어갔지만, 세 정점 연결정보가 한곳에 다 정렬돼있으므로 kruskal()을 이용한 것이 조금 더 동작이 빠르다.
