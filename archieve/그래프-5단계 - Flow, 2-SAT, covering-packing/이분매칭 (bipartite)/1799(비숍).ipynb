### 풀이
- 대각선 처리를 너무나도 못해서 답지를 봐버렸다..
  - i-j+N-1 까진 유도했지만 그 반대편이 j-i+N-1이라고 생각한게 화근이였다. 
- 9525(룩 처리하기) 와 같은 원리로 풀면 된다. 
  - 입력으로 0이 된 부분이 실제 비숍의 경로를 막지 않으므로, 각 좌표가 주워졌을 때 이것이 몇번 간선과 연결돼있는지 $O(1)$ 만에 알 수 있다는게 아이디어였다.
  - 대각선으로 처리할 때 `i+j`와 `j-i+(N; 그리드 길이)-1`의 관계는 기억해둘만 한것 같다.
    - 왼쪽 위부터 좌하단으로 (0, 0)에서 id가 0번, (4, 4)에서 id가 2N-1 번이게 설계하는 방법은 `i + j`
    - 오른쪽 위부터 우하단으로 (0, 4)에서 id가 0번, (4, 0)에서 id가 2N-1 번이게 설계하는 방법은 `j - i + (2N-1)` 번
