```python
import io, os
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def sol() :
  S = input().rstrip()
  cnt = {}
  is_full_S_pallindrome = True
  for i, s in enumerate(S) :
    if s != S[~i] :
      is_full_S_pallindrome = False
    cnt[s] = cnt.get(s, 0) + 1
  
  if len(cnt) == 1 :
    print(-1)
  elif is_full_S_pallindrome :
    print(len(S) - 1)
  else :
    print(len(S))
sol()
```

### 자력솔 및 증명
- 풀이시간: 약 11분
- N이 50만인 문제이다. 문자열을 $O(\log N)$ 느낌으로 순회하는건 약간 안와닿으니 풀이는 $O(N)$ 이라고 가정할 수 있다.
- 원래 알고 있던 사실 한개가 있었다. 어떤 문자열이 단일 문자열로 되어있다면, 그 문자열은 어느 곳에서 자르더라도 회문이다. 살펴보니 예제 3번이 -1로 그러한 내용을 암시하고 있었다.
  - 우선, $S$ 의 원소를 집합으로 표기한 것을 $X$ 라고 해보자. $|X| = 1$ 일 경우, $S$ 는 반드시 회문이다.
  - 하지만 그렇지 않은 경우는 어떨까? $|X| > 1$ 인 경우를 생각해봤다.
    - 이 과정을 증명에 가깝게 생각하다가 이 아이디어가 맞을 수도 있겠다는 확신을 하고 구현을 한 뒤 AC를 받았다.
    - 증명. $|X| > 1$ 일 때 회문 $S$ 에서 문자열 한개를 제거한 $S'$ 는 반드시 회문이 아니다. \
    (내가 풀 때 생각한 방법은 아니지만, 이때 `S[::-1]` 과 `S` 가 같은지만 확인하면 증명이 끝난다)\
    우선 substring의 정의에 의해 임의의 문자 한개가 제거된 $S`$ 는 `S[:-1]` 혹은 `S[1:]` 이다.\
    편의상 $\forall s \in S \in X$ 인 집합 $X$ 크기를 구하는 함수를 $f(S)$ 라고 해보자. 그리고 lemma가 아니라고 가정해보자.\
    \
    $f(S) = 2$ 인 경우 \
    잘려나간 문자 $c$가 나머지 문자열들과 다른 유일한 문자, 즉 $S$ 에 대한 $f(S') = 1$ 일 경우, 애초에 반대쪽의 문자가 $c$ 여야 한다.\
    따라서 다른 것과 구분되는 문자가 1개 일 경우, $S$ 의 양 끝에 존재하지 않는다. \
    즉, 양 끝단이 아닌 임의의 위치 $i$ 에 $c$ 가 존재한다.\
    $S'$ 에서 제거했던 문자의 반대쪽을 제거한 부분을 $S''$ 라고 하자. $c$ 는 $S''$ 안에 존재한다.\
    이때 $S$ 에서 한 글자를 뺏다는 의미는, $|S|$ 의 길이가 짝수/홀수 여부가 뒤바뀌었다는 의미이다.\
    그런데도 $S''$ 가 팰린드롬이라는 의미는, 홀수/짝수의 여부와 상관없이 팰린드롬을 만족한다는 의미이다.\
    $S''$ 가 홀수/짝수에 관계 없이 팰린드롬을 만족하는 경우는 $f(S'') = 1$ 일 때만 성립한다.\
    이는 가정에 모순이다.\
    $\therefore f(S'') \ne 1$ 일 경우, $S'$ 는 팰린드롬이 아니다.\
    \
    동일한 방법이 $f(S) > 2$ 일때도 성립한다. \
    따라서 $f(S) > 1$ 일 때, $S'$ 는 팰린드롬이 아니다.
  - 회문인 문자열 $S$에서 한 문자열을 뺀 것이 반드시 회문이 아니라면, 가장 긴 회문이 아닌 부분 문자열은 세가지 경우로 좁혀지게 된다.
    - $f(S) = 1$ 인 경우: -1을 출력
    - $f(S) \ne 1$ 이면서 $S$ 의 가장 긴 회문의 길이가 $N$ 인 경우: $N-1$ 을 출력
    - $f(S) \ne 1$ 이면서 $S$ 의 가장 긴 회문의 길이가 $N$ 이 아닌 경우: $N$ 을 출력
