```python
import sys, itertools
input=sys.stdin.readline

def sol() :
  N = int(input()) 
  A,B,C,D,E,F = map(int, input().split())
  L = [A, B, C, D, E, F]
  if N == 1 :
    x = max(A,B,C,D,E,F)
    return print(sum([A,B,C,D,E,F]) - x)

  a = min(A,B,C,D,E,F)
  b = min(L[i] + L[j] for i in range(6) for j in range(i+1, 6) if i+j != 5)
  c = min([L[i] + L[j] + L[k] for i in range(6) for j in range(i+1, 6) for k in range(j+1, 6) if i + j != 5 and j + k != 5 and k + i != 5])

  print(a * ((N-2)*4*(N-1) + (N-2)**2) + b * (8 * N - 12) + c * 4)
  
sol()
```

- (자력) 인접한 면의 숫자 합이 각각 1,2,3개인 주사위 면들의 합이 최소가 되도록 적절히 주사위를 회전시킬 수 있다.
  - 보이는 면이 3개인 곳 : 4개 
  - 보이는 면이 2개인 곳 : $8N - 12$ 개
  - 보이는 면이 1개인 곳 : $5N^2 - 16N + 12$ 개
    - 다만 수식을 구하는 과정에서 노가다 대신 면의 넓이로 구하는 게 조금 더 쉬웠을 것 같다.
    - $4(N-2)(N-1) + (N-2)^2$
    - 
- 주사위 양쪽면의 합은 7이다. (인덱스 합은 5다)
  - 따라서 인덱스 합이 5가 아닌것의 조합을 구하면 된다.
