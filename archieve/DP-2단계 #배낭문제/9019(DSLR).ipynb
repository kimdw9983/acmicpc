```python
import collections, sys
input=sys.stdin.readline
T = int(input())
for _ in range(T) :
  A, B = map(int, input().split())
  NONE = -1
  START = 10002

  def D(n):
    return (2 * n) % 10000

  def S(n) :
    return (n - 1) % 10000

  def L(n) :
    n = str(n)
    n = ('0' * (4 - len(n)) if len(n) < 4 else '') + n
    return int(n[1:] + n[0])

  def R(n) :
    n = str(n)
    n = ('0' * (4 - len(n)) if len(n) < 4 else '') + n
    return int(n[-1] + n[:-1])

  def recon(vis) :
    X = B
    answer = []
    QUERY = "DSLR"
    while vis[X] != START :
      for i, nX in enumerate([D(vis[X]), S(vis[X]), L(vis[X]), R(vis[X])]) :
        if X != nX : continue
        answer.append(QUERY[i])
        break
      X = vis[X]
    
    print(''.join(answer[::-1]))

  def query(x, y) :
    vis = [NONE] * 10000
    Q = collections.deque([x])
    vis[x] = START
    
    while Q :
      x = Q.popleft()
      if x == y : return recon(vis)

      for nx in (D(x), S(x), L(x), R(x)) :
        if vis[nx] != NONE : continue
        Q.append(nx)
        vis[nx] = x

  query(A, B)
```

## 마지막 index를 저장하여 해를 역추적한 방법
- 아무래도 DSLR의 역연산을 구현한 방법이 더 빠른 것으로 보인다. 
  - 이 구현의 수행시간은 무려 `7352`ms나 됐다
- `for i, nX in enumerate([D(vis[X]), S(vis[X]), L(vis[X]), R(vis[X])]) :`에서 vis[X]부분을 X로 적어놓고 있었다. 이것 때문에 시간이 오래 걸렸다.
  - 둘의 관계를 정확히 이해했으면 여기서 시간을 덜 들였을까?
  - 혹은 이해하지 못했더라도 parameter로 vis[X]를 넘겨볼 생각을 좀 더 빨리했어야 했다..
- ~~`if nX not in vis` :에서 vis[X]부분을 X로 적어놓고 있었다. 이것 때문에 시간이 오래 걸렸다.~~
  - 이런 실수는 하지 않았다. 미친 Copilot이 내가 안한 실수도 적어내려 한다...

```python
import collections, sys
input=sys.stdin.readline
T = int(input())
for _ in range(T) :
  A, B = map(int, input().split())
  NONE = -1
  START = 10002

  def D(n):
    return (2 * n) % 10000

  def S(n) :
    return (n - 1) % 10000

  def L(n) :
    n = str(n)
    n = ('0' * (4 - len(n)) if len(n) < 4 else '') + n
    return int(n[1:] + n[0])

  def R(n) :
    n = str(n)
    n = ('0' * (4 - len(n)) if len(n) < 4 else '') + n
    return int(n[-1] + n[:-1])

  def recon(vis, com) :
    X = vis[B]
    answer = [com[B]]
    while X != START :
      answer.append(com[X])
      X = vis[X]
    print(''.join(answer[::-1]))

  def query(x, y) :
    vis = [NONE] * 10000
    com = [''] * 10000
    Q = collections.deque([x])
    vis[x] = START
    
    while Q :
      x = Q.popleft()
      if x == y : return recon(vis, com)

      for c, nx in zip("DSLR", [D(x), S(x), L(x), R(x)]) :
        if vis[nx] != NONE : continue
        Q.append(nx)
        vis[nx] = x
        com[nx] = c

  query(A, B)
```
