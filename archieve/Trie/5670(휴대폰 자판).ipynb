```python
import sys
input=sys.stdin.readline

class Trie:
  def __init__(self):
    self.r = {}

  def add(self, s):
    cur = self.r
    for c in s:
      cur = cur.setdefault(c, {})
    cur["_end_"] = True

  def __contains__(self, s):
    cur = self.r
    for c in s:
      if c not in cur:
        return False
      cur = cur[c]
    return "_end_" in cur

def DFS(node, l, res) :
  n = 0 #자식 노드의 수 len(node) that is not "_end_"
  for v in node :
    if v == '_end_' : continue
    n +=1
  
  should_increase = int('_end_' in node or n > 1) #자식의 노드 수가 2개 이상이거나, 자신이 단어일 경우
  if '_end_' in node : #사전에 추가됏던 단어면
    res.append(l) #답 추가

  for v in node :
    if v == '_end_' : continue 
    DFS(node[v], l+should_increase, res)

def sol(n) :
  T = Trie()
  for _ in range(n) :
    T.add(input().rstrip())
  
  res = []
  for v in T.r.values() :
    DFS(v, 1, res)

  answer = sum(res) / len(res)
  sys.stdout.write(format(round(answer, 2), ".2f") + "\n")

while True :
  try :
    sol(int(input()))

  except:
    break
```

- 트라이 구조를 직접 그려보면 이해가 쉽다. 

### 풀이
- Trie로 Tree를 순회하면서 `버튼을 눌러야 하는 횟수` $l$의 값을 전달시킨다. 다음과 같은 경우일 때 해당하는 조건 노드의 자식에 대해 $l$을 1만큼 증가시킨다.
  - 자식의 개수가 2개 이상일 경우
  - 자신이 단어일 경우

### 주의사항
- 1번의 `모듈이 단어의 첫 번째 글자를 추론하지는 않는다`를 만족하게 하기 위한 초기 조건에 주의한다.
  - (실수) DFS를 무조건 1로 시작하면 안된다. 루트에서 자식노드가 `1개일 때`와 `2개 이상일 때`의 값이 달라질 수 있기 때문이다.
    - `DFS(v, 1, res)`
  - 대신 root의 자식에 대해 1을 증가시킨 채로 시작한다. 
    - `for v in T.r.values() : DFS(v, 1, res)`
- 출력 조건에 주의한다. 정답을 소수점 둘째자리까지 출력해야 하는데, 반올림만 하면 안된다.
  - `sys.stdout.write(format(round(answer, 2), ".2f") + "\n")`
