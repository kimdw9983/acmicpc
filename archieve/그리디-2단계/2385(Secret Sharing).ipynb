정렬 기준을 반대로 바꾸기만 하면 다음과 같은 반례가 있다.
- 0 21 2
  - 출력: 2102
  - 정답: 2021

### 풀이
- 0으로 시작하지 않는 최소값을 찾아야한다.
  - 아이디어는 0으로 시작하지 않는 숫자들 뒤에 0을 붙이고, 기존의 '가능한 가장 긴 자릿수 + 1'(6)개가 될 때 까지 반복해서 붙인 값을 비교하는 것이다.
  - ex) 20 / 2 / 21 => 200200 / 202020 / 210210
  - 하지만 첫번째로 출력할 값을 제외하면 다른 기준으로 출력해야 한다.

```python
import sys
input()
L = [*input().split()]

VALID = False
for v in L :
  if v[0] == '0': continue
  VALID = True

L2 = []
for v in L :
  x = v + '0'
  i = 0
  l2 = []
  while i < 7 :
    if i > len(x)-1 :
      l2.append(x[i%len(x)])
    else :
      l2.append(x[i])
    i += 1
  L2.append(int(''.join(l2)))

L3 = [x for _, x in sorted(zip(L2, L), key=lambda pair: -pair[0])]
#0이 앞에오지 않는 가장 작은 수
minimum = -1
while L3 :
  x = L3.pop()
  if x[0] == '0' : continue
  minimum = x
  break

if minimum == -1 :
  sys.stdout.write('INVALID')
else :
  L2 = []
  for v in L :
    x = v
    i = 0
    l2 = []
    while i < 6 :
      if i > len(x)-1 :
        l2.append(x[i%len(x)])
      else :
        l2.append(x[i])
      i += 1
    L2.append(int(''.join(l2)))

  L4 = [x for _, x in sorted(zip(L2, L), key=lambda pair: -pair[0])]
  L4.remove(minimum)

  S = []
  S2 = []
  while L4 :
    x = L4.pop()
    if x[0] == '0' :
      S.append(x)
      continue

    S2.append(x)
  sys.stdout.write(''.join([minimum] + S + S2))
```

- 0으로 시작하지 않는 기준에서 가장 작은 숫자 - 0이 포함된 숫자 - 가장 작은 숫자 순으로 출력하는 코드

  - 0이 없는데 minimum을 맨 앞에 출력하는 경우도 틀린 정답을 유도할 수 있어서 뺏는데 여전히 틀렸다

### 잠정 포기
- 반례를 도저히 찾을 수가 없다. 

```python
import sys
input()
L = [*input().split()]

L2 = []
for v in L :
  x = v + '0'
  i = 0
  l2 = []
  while i < 15 :
    l2.append(x[i%len(x)])
    i += 1
  L2.append(''.join(l2))

L3 = [x for _, x in sorted(zip(L2, L), key=lambda pair: pair[0], reverse=True)]
minimum = -1 #0이 앞에오지 않는 가장 작은 수
has_zero = False
while L3 :
  x = L3.pop()
  if x[0] == '0' : 
    has_zero = True
    continue
  minimum = x
  break

if minimum == -1 :
  sys.stdout.write('INVALID')
else :
  L2 = []
  for v in L :
    x = v
    i = 0
    l2 = []
    while i < 15 :
      l2.append(x[i%len(x)])
      i += 1
    L2.append(''.join(l2))

  L4 = [x for _, x in sorted(zip(L2, L), key=lambda pair: pair[0])]
  if has_zero :
    L4.remove(minimum)
  # print(L2)

  S = []
  answer = [minimum] if has_zero else []
  sys.stdout.write(' '.join(answer + L4))
```
