```python
K, N = map(int, input().split())
L = [int(input()) for _ in range(K)]

for _ in range(N-K) :
  L.append(max(L))

L2 = []
for v in L :
  x = str(v)
  i = 0
  l2 = []
  while i < 10 :
    if i > len(x)-1 :
      l2.append(x[-1])
    else :
      l2.append(x[i])
    i += 1
  L2.append(int(''.join(l2)))

L3 = [x for _, x in sorted(zip(L2, L), key=lambda pair: ( -pair[0], len(str(pair[1]))))]

for v in L3 :
  print(v, end="")
```

### 반례
- 10, 100, 1004
  - 10/1004/100
- 3213, 321, 32132
  - 3213/32132/321
- 가장 긴 수 An와 가장 긴 수의 첫번째 숫자A1를 합친 숫자를 An|A1라고 할때, 
  - 그것보다 자릿수가 적은 숫자Xn도 An|A1과 길이가 같을때까지 Xn|Xn|Xn..이런식으로 합친다. 
    - 합치는 도중 Xn이 도중에 잘려도 상관없다.
  - An|A1과 합친 숫자 Xn|Xn|Xn을 비교하여 큰 숫자를 우선으로 오게 만든다.

```python
K, N = map(int, input().split())
L = [int(input()) for _ in range(K)]

for _ in range(N-K) :
  L.append(max(L))

L2 = []
for v in L :
  x = str(v)
  i = 0
  l2 = []
  while i < 10 :
    if i > len(x)-1 :
      l2.append(x[i%len(x)])
    else :
      l2.append(x[i])
    i += 1
  L2.append(int(''.join(l2)))

L3 = [x for _, x in sorted(zip(L2, L), key=lambda pair: ( -pair[0], len(str(pair[1]))))]

for v in L3 :
  print(v, end="")
```
