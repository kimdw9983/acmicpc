```python
import io, os
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def switch(G, x, y) :
  for dx, dy in (0, 0), (1, 0), (-1, 0), (0, 1), (0, -1) :
    nx = x + dx
    ny = y + dy
    if not(0 <= nx < 10 and 0 <= ny < 10) : continue
    G[nx][ny] ^= 1

def sol() :
  G = [[] for _ in range(1 << 10)]
  cnt = [0] * (1 << 10)
  for _ in range(10) :
    l = list(map(lambda x: int(x == 79), input().rstrip()))
    for bit in range(1 << 10) :
      G[bit].append(l[:])

  for bit in range(1 << 10) :
    c = 0
    for i in range(10) :
      if bit & (1 << i) :
        switch(G[bit], 0, i)
        c += 1
    cnt[bit] = c
  
  for x in range(1, 10) :
    for bit in range(1 << 10) :
      c = 0
      for y in range(10) :
        if G[bit][x-1][y] == 1 :
          switch(G[bit], x, y)
          c += 1
      cnt[bit] += c
  
  ans = 101
  for bit in range(1 << 10) :
    if all(G[bit][i][j] == 0 for i in range(10) for j in range(10)) :
      ans = min(ans, cnt[bit])
  
  if ans == 101 : return print(-1)
  print(ans)

sol()
```

### 자력솔(with tag hint)
- 2138(전구와 스위치)의 확장 문제.
  - 위 문제의 그리디 아이디어는 `n번째 전구를 건드릴 때 n-1번째 전구의 상태가 확정된다` 이다.
  - 이 문제에도 동일한 아이디어를 사용할 수 있다. `현재 줄의 상태를 건드릴 때, n-1번째 줄의 상태가 확정된다.`
    - 단, 현재 줄에 대해서 어떻게 눌러야 할지는 확정할 수 없으므로(어떤 칸을 누르지 않아야 될 수도 있고, 눌러야만 할 수도 있다) 전부 수행해봐야 한다. \
    10개의 버튼을 각각 눌러보는 경우의 수는 $2^{10}$
- 시간복잡도 $\Omicron(2 ^ {10} * 10 * 10)$
- 2^10 * @ 정도는 할 수 있을 것이라는 발상이 됐으면 태그를 보지 않았더라도 풀 수 있었을 지도 모른다.
