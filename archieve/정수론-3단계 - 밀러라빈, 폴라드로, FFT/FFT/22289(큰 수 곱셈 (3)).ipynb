```python
import sys
input = sys.stdin.readline

p = 469762049
def ntt(a, inv=False):
  n = len(a)
  j = 0
  for i in range(1,n):
    rev = n >> 1
    while j >= rev:
      j -= rev
      rev >>= 1
    j += rev
    if i < j:
      a[i], a[j] = a[j], a[i]

  step = 2
  while step <= n:
    half = step // 2
    u = pow(3,p//step,p)
    if inv: u = pow(u,p-2,p)
    for i in range(0, n, step):
      w = 1
      for j in range(i, i + half):
        v = a[j + half] * w
        a[j + half] = (a[j] - v)% p
        a[j] += v
        a[j] %= p
        w *= u
        w %= p
    step *= 2

  if inv:
    invn = p - (p-1) // n
    for i in range(n):
      a[i] = (a[i] * invn)%p
  return a
    
def ntt_conv(a, b):
  s = len(a) + len(b) - 1
  n = 1 << s.bit_length()
  a += [0 for _ in range(n - len(a))]
  b += [0 for _ in range(n - len(b))]
  ntt(a)
  ntt(b)
  for i in range(n):
    a[i] *= b[i]
  ntt(a,True)
  return a

def sol() :
  A, B = input().split()
  if A == '0' or B == '0' : return print(0)
  P = []
  Q = []
  num0A = len(A) - len(A.rstrip('0'))
  num0B = len(B) - len(B.rstrip('0'))

  for v in A:
    P.append(int(v))
  
  for v in B:
    Q.append(int(v))
  
  R = ntt_conv(P, Q)
  while R[-1] == 0 :
    R.pop()
  
  mem = 0
  res = []
  for _ in range(num0A + num0B) : #ntt는 0을 처리할 수 없어보여서 0을 따로 처리했다.
    res.append('0')

  for n in reversed(R) : #흔히 어렸을때 배웠던 곱셈방식. 각 자릿수의 곱과 올림수를 처리한다.
    tmp = n + mem
    res.append(str(tmp % 10))
    mem = tmp // 10

  if mem :
    res.append(str(mem))

  res.reverse()
  print("".join(res))

sol()
```

### 자력솔
- 입력된 숫자에 대해서 각 자릿수를 $10^k$ 에 계수를 곱한 것으로 생각할 수 있고, 이를 통해 다항식을 만든 뒤 곱셈을 하면 된다.
- 곱셈을 하면 각 원소에 10이 넘는 숫자가 들어있는데, 어렸을 때 곱셈을 배울 때 올림을 하면서 계산했던 것처럼, 이를 통해 다시 계수를 구하면 된다.
- 또한 곱할 수가 0으로 끝나는 경우, ntt와 ftt는 배열의 사이즈를 2의 승수로 맞추느라 랜덤한 만큼 0을 채워 넣기 때문에 올바른 정보를 얻을 수 없다.
  - 따라서 처음에 각 숫자의 0의 개수를 센뒤, 곱셈을 하고 나서 0의 개수만큼 뒤에 0을 채워 넣어준다.
