```python
import io, os
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

INF = float("inf")
class Dinic:
  def __init__(self, n):
    self.lvl = [0] * n
    self.ptr = [0] * n
    self.q = [0] * n
    self.adj = [[] for _ in range(n)]

  def add(self, u, v, c, rcap=0):
    self.adj[u].append([v, len(self.adj[v]), c, 0])
    self.adj[v].append([u, len(self.adj[u]) - 1, rcap, 0])

  def dfs(self, v, t, f):
    if v == t or not f:
      return f

    for i in range(self.ptr[v], len(self.adj[v])):
      e = self.adj[v][i]
      if self.lvl[e[0]] == self.lvl[v] + 1:
        p = self.dfs(e[0], t, min(f, e[2] - e[3]))
        if p:
          self.adj[v][i][3] += p
          self.adj[e[0]][e[1]][3] -= p
          return p
      self.ptr[v] += 1

    return 0

  def calc(self, s, t):
    flow, self.q[0] = 0, s
    for l in range(31):
      while True:
        self.lvl, self.ptr = [0] * len(self.q), [0] * len(self.q)
        qi, qe, self.lvl[s] = 0, 1, 1
        while qi < qe and not self.lvl[t]:
          v = self.q[qi]
          qi += 1
          for e in self.adj[v]:
            if not self.lvl[e[0]] and (e[2] - e[3]) >> (30 - l):
              self.q[qe] = e[0]
              qe += 1
              self.lvl[e[0]] = self.lvl[v] + 1

        p = self.dfs(s, t, INF)
        while p:
          flow += p
          p = self.dfs(s, t, INF)

        if not self.lvl[t]:
          break

    return flow

def sol():
  N, K, D = map(int, input().split())
  src, snk = 0, N + D + 1
  F = Dinic(N + D + 2)
  for i in range(N) :
    F.add(src, i+1, K)

  L = list(map(int, input().split()))
  for i, v in enumerate(L) :
    F.add(N+i+1, snk, v)
  
  for i in range(N) :
    _, *L = map(int, input().split())
    for j in L :
      F.add(i+1, N+j, 1)

  ans = F.calc(src, snk)
  return print(ans)

sol()
```
