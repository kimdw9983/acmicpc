```python
import sys
input=sys.stdin.readline

from heapq import heappush, heappop

def sol() :
  N, M = map(int, input().split())
  G = [[*map(int, input().split())] for _ in range(N)]
  if G[0][0] == -1 : return print(-1)

  hq = [(G[0][0], 0, 0)]
  vis = [[False] * M for _ in range(N)]
  vis[0][0] = True

  while hq :
    c, x, y = heappop(hq)
    if (x, y) == (N-1, M-1) :
      return print(c)
    
    for dx, dy in (0, 1), (0, -1), (-1, 0), (1, 0) :
      nx, ny = x + dx, y + dy
      if not (0 <= nx < N and 0 <= ny < M) : continue
      if G[nx][ny] == -1 : continue
      if vis[nx][ny]: continue

      vis[nx][ny] = True
      heappush(hq, (c + G[nx][ny], nx, ny))
  
  return print(-1)

sol()
```

- 다익스트라가 이렇게 확장될 수 있음을 배웠다.
- heapq는 첫번째 원소를 기준으로 정렬함에 주의. x,y,c순이 아닌 c,x,y 순으로 해야한다.
