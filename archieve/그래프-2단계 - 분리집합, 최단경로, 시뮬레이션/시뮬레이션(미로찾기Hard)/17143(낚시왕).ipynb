```python
import io, os, collections
input=io.BytesIO(os.read(0, os.fstat(0).st_size)).readline


def sol() :
  R, C, M = map(int, input().split())
  G = [[None] * C for _ in range(R)]
  nextG = [[[] for _ in range(C)] for _ in range(R)]

  class Shark :
    def __init__(self, s, d, z) :
      self.moved = False
      self.speed = s
      self.dir: tuple[int, int] = self.calc_dir(d)
      self.size = z

    @staticmethod
    def calc_dir(d) :
      if d == 1 : return (-1, 0)
      elif d == 2 : return (1, 0)
      elif d == 3 : return (0, 1)
      elif d == 4 : return (0, -1)

    def move(self, r, c) :
      left  = self.speed #남은 이동량
      dir = self.dir
      while left > 0 :
        nr, nc = r + dir[0], c + dir[1]
        if (0 <= nr < R and 0 <= nc < C) :
          r = nr
          c = nc
        else :
          dir = (dir[0] * -1, dir[1] * -1)
          r += dir[0]
          c += dir[1]
        left -= 1

      self.dir = dir
      self.moved = True
      return r, c

    def __repr__(self): 
      return f"({self.size},{self.dir},{self.speed})"

  for _ in range(M) :
    r, c, s, d, z = map(int, input().split())
    G[r-1][c-1] = Shark(s, d, z)

  def move_eat() :
    for i in range(R) :
      for j in range(C) :
        while G[i][j] :
          shark: Shark = G[i][j]
          r, c = shark.move(i, j)
          nextG[r][c].append(shark)
          G[i][j] = None

    for i in range(R) :
      for j in range(C) :
        if nextG[i][j] :
          apex: Shark = nextG[i][j].pop()
          while nextG[i][j] :
            shark: Shark = nextG[i][j].pop()
            if shark.size > apex.size :
              apex = shark
          G[i][j] = apex
  
  ans = 0
  for i in range(C) :
    for j in range(R) :
      if G[j][i] :
        ans += G[j][i].size
        G[j][i] = False
        break
    move_eat()
  
  print(ans)

sol()
```

### 자력솔
- 문제에서 요구한걸 그대로 시뮬레이션 하면 된다.
- 상어는 size, speed, direction을 가지고 있고, 그것의 이동량은 전역범위에서 관리하는 발상을 조금 더 빨리했으면 좋았을 것 같다.
