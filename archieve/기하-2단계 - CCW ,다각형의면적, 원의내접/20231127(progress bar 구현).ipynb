### 각도 단위(JavaScript)
- `deg`: 도. 직관적인 기준점. $360 \degree$ 는 한바퀴이다.
- `rad`: 라디안. $2\pi \text{ rad}  = 360\degree$
- `grad`: 그라드. $400 \text{ grad} = 360\degree$
- `turn`: 턴. $1 \text{ turn} = 360\degree$
- `%`: 퍼센트. $100\% = 360\degree$

### conic-gradient
- 
- conic-gradient() 함수는 원뿔 모양의 그라데이션을 생성한다.
- conic-gradient(c1 t1, c2 t2, c3 t3, ...)
  - 360도의 원을 그리면서 그 비율에 대한 색깔 $c_i$ 를 기준점으로부터 중심각이 $t_i$ 인 기준 호를 경계로 그라디언트를 채운다.
  - $t_i$ 는 처음과 끝의 경계를 지정해줄 수도 있다.
    - conic-gradient(red 36deg, orange 36deg 170deg, yellow 170deg) 의 결과\
    
    - 이 테크닉을 이용하면 흔히 말하는 그라디에이션이 아닌, 딱딱한 경계를 가진 원을 그릴 수 있게된다.
  - 초기 각도와 방향은 상대위치 50%, 50%를 중점으로 하여, 0도부터 시작한다.
  - 중점과 초기 각도는 `from` 과 `at` 키워드로 변경할 수 있다.
    - 여기서 다루진 않았지만, 잘 다룬다면 그라디언트로 체크판을 만들수도 있다..
    - 
- conic-gradient(c1, c2, c3, ...)
  - 위와 비슷하지만, $c_i$ 색깔을 가지는 기준 호의 크기는 모두 동일하다.

### 유도
- 0일때 (50, -2.25), 25% 일 때 (102.5, 50), 50%일 때 (50, 102.5), 75%일 때 (-2.25 50)
- 중점이 (50, 50)이고, 반지름이 52.5인 원이라고 해석할 수 있다.
- 입력 값의 100%가, 중심각을 기준으로 75%인 호를 그려야 한다.\
즉, 입력 범위인 [0, 1] 을 [0, 0.75] 로 normalize 해야 한다는 의미.
