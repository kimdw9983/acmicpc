```python
import math
def get_2dline(p1, p2):
    if p1 == p2:
        return (0, 0, 0)
    _p1, _p2 = min(p1, p2), max(p1, p2)
    a, b, c = _p2[1] - _p1[1], _p1[0] - _p2[0], _p1[1] * _p2[0] - _p1[0] * _p2[1]
    g = math.gcd(math.gcd(a, b), c)
    return (a // g, b // g, c // g)

is_parallel = lambda l1, l2: l1[0] * l2[1] == l2[0] * l1[1]
is_same = lambda l1, l2: is_parallel(l1, l2) and (l1[1] * l2[2] == l2[1] * l1[2])
collinear = lambda p1, p2, p3: is_same(get_2dline(p1, p2), get_2dline(p2, p3))

x1, y1, x2, y2, x3, y3 = map(int, input().split())
if collinear((x1, y1), (x2, y2), (x3, y3)):
  print(float(-1))
else :
  d1 = math.dist((x1, y1), (x2, y2))
  d2 = math.dist((x1, y1), (x3, y3))
  d3 = math.dist((x2, y2), (x3, y3))

  a = 2 * min(d1, d2, d3)
  b = 2 * max(d1, d2, d3)
  print(abs(a-b))
```

### 자력솔
- 세 점이 평행하면 평행사변형을 만들 수 없다.
- 
- d1, d2, d3의 길이를 이용해 평행사변형을 만들어보고 그 최대값과 최소값의 차이를 출력한 것이 정답이 된다.
