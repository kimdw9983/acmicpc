```python
from heapq import heappop, heappush
import sys
input=sys.stdin.readline

def sol():
  V, E, T = map(int, input().split())
  s, g, h = map(int, input().split())
  G = [[] for _ in range(V+1)]
  for _ in range(E) :
    u, v, w = map(int, input().split())
    G[u].append((v, w))
    G[v].append((u, w))

  l = []
  for _ in range(T) :
    l.append(int(input()))

  def dij(G, s):
    n = len(G)
    dist, parents = [float("inf")] * n, [-1] * n
    dist[s] = 0

    Q = [(0, s)]
    while Q:
      pl, v = heappop(Q)
      if pl != dist[v]: continue
      for w, el in G[v]:
        if el + pl >= dist[w]: continue
        dist[w], parents[w] = el + pl, v
        heappush(Q, (el + pl, w))

    return dist, parents

  d1, p1 = dij(G, s)
  d2, p2 = dij(G, g)
  d3, p3 = dij(G, h)
  can = [(0, 0)] #거리가 가장 짧은 index
  min = float("inf")
  for i, v in enumerate(d2):  #i = 정점, v = 거리
    if not (v and i in l): continue #자기 자신이거나 정답 후보에 없으면 continue
    if min > d1[h] + v: 
      can = [(i, p2[i])]
      min = d1[h] + d2[can[-1][0]]
    elif min == d1[h] + v: can.append((i, p2[i]))

  for i, v in enumerate(d3) :
    if not (v and i in l): continue
    if min > d1[g] + v: 
      can = [(i, p3[i])]
      min = d1[g] + d3[can[-1][0]]
    elif min == d1[g] + v: can.append((i, p3[i]))

  answer = set()
  for i, parent in can:
    for j in G[parent] :
      if j[0] in l :
        answer.add(j[0])
  
  print(" ".join(map(str, sorted(answer))))
  
for _ in range(int(input())):
  sol()
```

- "최단거리로 갈 것"이라는 의미를 잘못 이해함.
- 최단거리로 이동한 경로에, 반드시 지나가는 경로가 포함되지 않은 후보를 제외하는 문제였다.
  1. '반드시 경로'를 통과하는 최단거리와 그냥 최단거리를 구한다.
  2. '반드시 경로'의 최소값이 아닌 후보들을 제외한 정점을 출력한다. 

```python
from heapq import heappop, heappush
import sys
input=sys.stdin.readline

def sol():
  V, E, T = map(int, input().split())
  s, g, h = map(int, input().split())
  G = [[] for _ in range(V+1)]
  for _ in range(E) :
    u, v, w = map(int, input().split())
    G[u].append((v, w))
    G[v].append((u, w))

  l = []
  for _ in range(T) :
    l.append(int(input()))
  l.sort()

  def dij(G, s):
    n = len(G)
    dist = [123456791] * n
    dist[s] = 0

    Q = [(0, s)]
    while Q:
      pl, v = heappop(Q)
      if pl != dist[v]: continue
      for w, el in G[v]:
        if el + pl >= dist[w]: continue
        dist[w] = el + pl
        heappush(Q, (el + pl, w))

    return dist

  for i, d in G[g]:
    if i != h: continue
    gh = d
    break

  d1 = dij(G, s)
  d2 = dij(G, g)
  d3 = dij(G, h)

  answer = []
  for v in l:
    if not (d1[g] + gh + d3[v] == d1[v] or d1[h] + gh + d2[v] == d1[v]): continue 
    answer.append(v)
  print(" ".join(map(str, answer)))
  
for _ in range(int(input())):
  sol()
```

- 좌변과 우변이 모두 `inf`일수도 있기 때문에 INF값을 `float("inf")`로 설정하는 것은 피해야한다.
