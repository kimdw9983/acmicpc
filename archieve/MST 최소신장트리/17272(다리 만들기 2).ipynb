```python
N, M = map(int, input().split())
G = [list(map(int, input().split())) for _ in range(N)]
D = [[0]*M for _ in range(N)] #섬의 번호

L= []# 다리가 이어질 곳의 찾을 섬 땅의 후보 땅의 좌표
def bfs(x, y, cnt):
  Q = [(x, y)]
  L.append((x, y))
  D[x][y] = cnt
  while Q:
    x, y = Q.pop()
    for dx, dy in (0, 1), (0, -1), (1, 0), (-1, 0):
      nx, ny = x + dx, y + dy
      if not (0 <= nx < N and 0 <= ny < M): continue
      if G[nx][ny] == 0 or D[nx][ny] > 0: continue
      D[nx][ny] = cnt
      Q.append((nx, ny))
      L.append((nx, ny))

cnt = 1
for i, l in enumerate(G) :
  for j, v in enumerate(l) :
    if v and not D[i][j] :
      bfs(i, j, cnt)
      cnt += 1
total_cnt = cnt - 1

V = []
E = []
W = []
while L : #후보 땅의 좌표에서 후보 다리를 이어보기
  x, y = L.pop()
  island_num = D[x][y] #현재 땅의 좌표
  for dx, dy in (0, 1), (0, -1), (1, 0), (-1, 0):
    nx, ny = x, y
    cnt = 0 #다리의 길이
    while True :
      nx, ny = nx + dx, ny + dy
      if not (0 <= nx < N and 0 <= ny < M): break
      if D[nx][ny] : #땅에 닿으면
        if D[nx][ny] == island_num : break #같은 땅에 닿으면 제외
        if cnt < 2 : break#길이가 2보다 작으면 제외
        V.append((x, y))
        E.append((nx, ny))
        W.append(cnt)
        break
      cnt += 1

class UF:
  def __init__(self, n):
    self.p = list(range(n))

  def find(self, a):
    a_ = a
    while a != self.p[a]:
      a = self.p[a]
    while a_ != a:
      self.p[a_], a_ = a, self.p[a_]
    return a

  def merge(self, a, b):
    self.p[self.find(b)] = self.find(a)

def kruskal(n, U, V, W): 
  uni = UF(n)
  c = 0
  L = sorted(range(len(W)), key=lambda x: W[x])
  for i in range(len(W)):
    ux, uy = U[L[i]]
    vx, vy = V[L[i]]
    u, v = D[ux][uy], D[vx][vy]
    fu, fv = uni.find(u), uni.find(v)
    if fu != fv:
      c += W[L[i]]
      uni.p[fv] = fu

  return c, uni

answer, uni = kruskal(total_cnt + 1, V, E, W)
if len(set(uni.find(i) for i in range(total_cnt + 1))) == 2 :
  print(answer)
else :
  print(-1)
```

## 풀이
- 풀이를 떠올리는데에 약 1시간 반정도 걸린 것 같다.
1. 그래프 탐색을 통해 섬의 개수와 섬의 번호를 붙인다
2. 섬의 땅에서 그래프 탐색을 통해 다음과 같은 조건의 후보 다리를 만든다.
    - 거리가 2 이상일 것
    - 경계를 벗어나지 않을 것
    - 자기 섬의 땅이 아닐 것(ex 도넛모양)
3. 다리 후보들을 출발 정점, 도착 정점, 거리의 list로 만든다.
4. kruskal 알고리즘을 통해 최소 신장 트리를 구한다.
    - 이때 사용된 union-find에 의해 섬끼리의 경로가 만들어지면 다리를 자연스럽게 만들지 않게 된다.
5. union list를 이용해 모든 섬이 이어졌는지 확인한다
    - 각각의 원소에 find를 적용하여 같은 부모를 가르키는지 확인한다.`len(set(uni.find(i) for i in range(total_cnt + 1)))`
        - 이때, union list 안에 있는 길이만을 검사`len(set(uni.p))`하면 안된다.
        - 직계 부모만 적혀있기 때문, 즉 원소 하나하나당 find를 적용시켜야 한다.
        - 예시를 만들 때 깊이가 여러개로 될 수 있는 경우를 만들지 못했다. 어떻게 했어야 했을까..
