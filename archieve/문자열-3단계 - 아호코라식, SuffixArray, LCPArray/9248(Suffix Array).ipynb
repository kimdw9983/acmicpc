```python
import io, os, sys
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def SAIS(S):
  n = len(S)
  buckets = [0] * (max(S) + 2)
  for a in S:
    buckets[a + 1] += 1
  for b in range(1, len(buckets)):
    buckets[b] += buckets[b - 1]
  isL = [1] * n
  for i in reversed(range(n - 1)):
    isL[i] = +(S[i] > S[i + 1]) if S[i] != S[i + 1] else isL[i + 1]

  def induced_sort(LMS):
    SA = [-1] * (n)
    SA.append(n)
    endpoint = buckets[1:]
    for j in reversed(LMS):
      endpoint[S[j]] -= 1
      SA[endpoint[S[j]]] = j
    startpoint = buckets[:-1]
    for i in range(-1, n):
      j = SA[i] - 1
      if j >= 0 and isL[j]:
        SA[startpoint[S[j]]] = j
        startpoint[S[j]] += 1
    SA.pop()
    endpoint = buckets[1:]
    for i in reversed(range(n)):
      j = SA[i] - 1
      if j >= 0 and not isL[j]:
        endpoint[S[j]] -= 1
        SA[endpoint[S[j]]] = j
    return SA

  isLMS = [+(i and isL[i - 1] and not isL[i]) for i in range(n)]
  isLMS.append(1)
  LMS = [i for i in range(n) if isLMS[i]]
  if len(LMS) > 1:
    SA = induced_sort(LMS)
    LMS2 = [i for i in SA if isLMS[i]]
    prev = -1
    j = 0
    for i in LMS2:
      i1 = prev
      i2 = i
      while prev >= 0 and S[i1] == S[i2]:
        i1 += 1
        i2 += 1
        if isLMS[i1] or isLMS[i2]:
          j -= isLMS[i1] and isLMS[i2]
          break
      j += 1
      prev = i
      SA[i] = j
    LMS = [LMS[i] for i in SAIS([SA[i] for i in LMS])]
  return induced_sort(LMS)

def KASAI(S, SA):
  n = len(S)
  rank = [0] * n
  for i in range(n):
    rank[SA[i]] = i
  LCP = [0] * (n - 1)
  k = 0
  for i in range(n):
    SAind = rank[i]
    if SAind == n - 1:
      continue
    j = SA[SAind + 1]
    while i + k < n and S[i + k] == S[j + k]:
      k += 1
    LCP[SAind] = k
    k -= k > 0
  return LCP

def sol() :
  S1 = input().rstrip()
  S2 = input().rstrip()

  concat = S1 + b"1" + S2
  SA = SAIS(concat)
  LCP = KASAI(concat, SA)

  idx, lcs = 0, 0 
  for i, lcp in enumerate(LCP) :
    if lcs < lcp and (SA[i] < len(S1)) != (SA[i+1] < len(S1)): 
      # 두 문자열의 common substring은 사전순으로 인접해있을 것이다. 
      # lcs의 최댓값을 갱신할 수 있을 때, 인접한 두 문자열 중 하나는 S1에 속해야 하고, 다른 하나는 S2에 속해야 한다.
      idx, lcs = i, lcp

  sys.stdout.write(f"{str(lcs)}\n{str(concat[SA[idx]:SA[idx]+lcs], 'ascii')}")
sol()
```

### 풀이
- 두 문자열의 common substring은 사전순으로 인접해있을 것이다. 즉, suffix array 상에서도 인접해있다.\
따라서 지금까지 구한 lcs보다 lcp값이 더 크고, 인접한 두 suffix가 서로 다른 문자열에 속한 것이라면, lcs의 최댓값을 갱신한다.
- 만약 LCP의 최댓값만 본다면 common substring이 아닌 두 문자열 중에 공통접두사가 더 긴 쪽을 출력하게 될 것이다.
  - banananananana / ana
- S1과 S2를 붙일 때 더미문자 하나를 붙여야 한다. 그렇지 않으면 둘중에 나머지 하나의 접두사인 경우, 더 긴쪽을 출력하기 때문이다.
  - bbaa / bbaabb
