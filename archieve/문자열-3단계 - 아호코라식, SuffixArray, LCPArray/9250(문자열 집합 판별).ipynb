```python
import io, os, sys
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

from collections import deque 
class AhoCorasick:
  def __init__(self, P, size=10 ** 5, c_size=26, f=lambda x: x - 97): # 전처리 O( sum(len(p) for p in P ) )
    self.trie = [[0] * c_size for _ in range(size)] 
    self.fail = [0] * size 
    self.res = [False] * size # output check
    self.f = f

    id = 0
    for p in P:
      v = 0
      for c in p :
        i = f(c)
        if not self.trie[v][i] :
          id += 1
          self.trie[v][i] = id
        v = self.trie[v][i]
      self.res[v] = True
    
    Q = deque([self.trie[0][i] for i in range(c_size) if self.trie[0][i]])
    while Q : # O( size * c_size )
      v = Q.popleft()
      for i in range(c_size) :
        if not self.trie[v][i] : continue #현재 상태에서 문자열 매칭이 있을 경우
        u = self.fail[v]
        while u and not self.trie[u][i] : #실패 함수를 따라가면서 문자열 매칭이 있는 상태를 찾는다. 없으면 루트로 돌아간다.
          u = self.fail[u]
        u = self.trie[u][i]
        self.fail[self.trie[v][i]] = u
        self.res[self.trie[v][i]] |= self.res[u]
        Q.append(self.trie[v][i])

  def __contains__(self, s): # O( len(s) )
    v = 0
    for c in s :
      i = self.f(c)
      while v and not self.trie[v][i] : 
        v = self.fail[v]
      v = self.trie[v][i]
      if self.res[v] : return True
    return False

def sol() :
  N = int(input())
  P = [input().rstrip() for _ in range(N)]
  Q = int(input())
  S = [input().rstrip() for _ in range(Q)]
  ac = AhoCorasick(P)

  ans = []
  for s in S:
    ans.append("YES" if s in ac else "NO")
  
  sys.stdout.write("\n".join(ans))
  
sol()
```

- 아호코라식 기본 구현 문제

```python
import io, os, sys
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

C_SIZE = 26
C_TO_I = lambda x: x - 97
class Trie:
  def __init__(self) :
    self.go = [None] * C_SIZE
    self.fail = None # 실패시 돌아갈 노드
    self.output = False

  def add(self, k) :
    if not k :
      self.output = True
      return

    v = C_TO_I(k[0])
    if self.go[v] is None :
      self.go[v] = Trie()
    self.go[v].add(k[1:])

from collections import deque 
class AhoCorasick:
  def __init__(self, trie: Trie): # 전처리 O( sum(len(p) for p in P ) )
    self.trie = trie
    self.trie.fail = self.trie
    
    Q = deque([self.trie]) 
    while Q :
      u = Q.popleft()
      for i in range(C_SIZE) :
        v = u.go[i]
        if not v: continue
        
        if u == self.trie :
          v.fail = self.trie
        else :
          w = u.fail
          while w != self.trie and not w.go[i] :
            w = w.fail
          if w.go[i] :
            w = w.go[i]
          v.fail = w
        v.output |= v.fail.output
        Q.append(v)
      
  def __contains__(self, s: str): # O( len(s) )
    u = self.trie
    for c in s :
      i = C_TO_I(c)
      while u != self.trie and not u.go[i] :
        u = u.fail
      if u.go[i] :
        u = u.go[i]
      if u.output :  #순회중에 가능한 문자열을 찾았으면 반환.
        return True 
    return False

def sol() :
  N = int(input())
  trie = Trie()
  for _ in range(N) :
    trie.add(input().rstrip())

  Q = int(input())
  S = [input().rstrip() for _ in range(Q)]
  ac = AhoCorasick(trie)

  ans = []
  for s in S:
    ans.append("YES" if s in ac else "NO")
  
  sys.stdout.write("\n".join(ans))
  
sol()
```
