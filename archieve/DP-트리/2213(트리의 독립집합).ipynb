```python
import io, os, sys
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def sol() :
  N = int(input())
  L = [*map(int, input().split())] #가중치
  G = [[] for _ in range(N)]
  for _ in range(N-1):
    a, b = map(int, input().split())
    G[a-1].append(b-1)
    G[b-1].append(a-1)

  DP = [[[0, 0] for _ in range(2)] for _ in range(N)] 
  #DP[i][j] = i번째 노드까지의 트리중 현재까지 만든 가중치의 최대값, j=0이면 i번째 노드를 포함한 경우, j=1이면 i번째 노드를 포함하지 않은 경우
  #DP[i][j] = [가중치 최대값, 자손의 노드의 값을 포함했는지의 여부]
  S = [0]
  vis = [False] * N
  PO = []
  while S:
    cur = S[-1]
    vis[cur] = True
    leaf = True
    for n in G[cur] :
      if not vis[n] :
        S.append(n)
        leaf = False
    
    if leaf :
      PO.append(S.pop())

  for cur in PO:
    for p in G[cur] :
      if vis[p] :
        # DP[cur][0][0] += L[cur] #내꺼를 우선 더함.
        # DP[cur][0][1] = 0 #항상 포함하지 않는다.
        DP[p][0][0] = L[p] + DP[cur][1][0]
        DP[p][0][1] = 0 #항상 포함하지 않는다.

        if DP[cur][1][0] > DP[cur][0][0] : #자손을 포함한 경우가 더 큰 경우
          DP[p][1][0] = DP[cur][1][0]
          DP[p][1][1] = 0 # 포함함
        else :
          DP[p][1][0] = DP[cur][0][0]
          DP[p][1][1] = 1 #포함하지 않음

    vis[cur] = False
  
  print(DP)

sol()
```

```python
import io, os, sys
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def sol() :
  N = int(input())
  L = [*map(int, input().split())] #가중치
  G = [[] for _ in range(N)]
  for _ in range(N-1):
    a, b = map(int, input().split())
    G[a-1].append(b-1)
    G[b-1].append(a-1)

  DP = [[0] * 2 for _ in range(N)]
  recon = [[[] for _ in range(2)] for _ in range(N)]
  #DP[i][j] = i번째 노드까지의 트리중 현재까지 만든 가중치의 최대값, j=0이면 i번째 노드를 포함 안한경우, j=1이면 i번째 노드를 포함한 경우
  S = [0]
  vis = [False] * N
  PO = []
  while S:
    cur = S[-1]
    vis[cur] = True
    leaf = True
    for n in G[cur] :
      if not vis[n] :
        S.append(n)
        leaf = False
    
    if leaf :
      PO.append(S.pop())

  for cur in PO:
    DP[cur][1] += L[cur]
    recon[cur][1].append(cur+1)
    for d in G[cur] :
      if vis[d] : #d는 부모 index
        #자신을 포함할 경우
        DP[d][1] += DP[cur][0] #자식을 포함하지 않은 경우의 가중치의 합을 더함
        for v in recon[cur][0] : #자식을 안넣은 경우의 index를 추가
          recon[d][1].append(v)

        #포함하지 않은 경우
        index = 0 #넣는 것과 안넣는 것중 큰 쪽의 index
        if DP[cur][1] > DP[cur][0] :
          index = 1 

        DP[d][0] += DP[cur][index]
        for v in recon[cur][index] :
          recon[d][0].append(v)
    vis[cur] = False
    
  index = 0
  if DP[0][1] > DP[0][0] :
    index = 1
  
  print(DP[0][index])
  sys.stdout.write(" ".join(map(str, sorted(recon[0][index]))))

sol()
```

- 후위순위를 순회할 때 `if vis[d]:` 이후가 부모 노드인데 자식노드인걸로 착각했다. 
  - iterative한 구현 예시가 없어서 약간 삽질
- DP의 기저상태를 제대로 설정하지 못했다.
