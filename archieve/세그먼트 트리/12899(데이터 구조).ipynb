### 접근
- 구하려는 수열의 길이 최대가 200만이므로, 이걸로 세그트리를 만들어도 충분할 것 같다. (약 30.5MB)
- `L[i] = 수열 안에 i가 있는 횟수` + 누적합 세그
- 쿼리 1는 `L[X]`의 값을 1 증가시키면 될 것이다.
- 쿼리 2는 root부터 자식을 살피면서 `X`번째로 작은 숫자를 찾으면 될 것 같다.
  - 좌우 자식을 살펴보면서 좌측 자식의 값보다 `X`가 크다면 오른쪽 자식으로 간다. 그렇지 않을경우 왼쪽으로 간다.
  - N번째보다 큰 숫자라는걸 누적하는 값`acc`를 재귀문에 넣는다. 
  - 오른쪽 자식으로 간다면 왼쪽 자식의 값을 `acc`에 누적시킨다.
  - leaf 노드를 찾았다면 그 노드의 인덱스인 i를 1 감소하고, i 값을 반환한다. 

```python
import io, os, sys
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

class ST:
  def __init__(self, L, default, f):
    self._def = default
    self._f = f
    self._len = len(L)
    self._size = _size = 1 << (self._len - 1).bit_length()

    self.L = [default] * (2 * _size)
    self.L[_size:_size + self._len] = L
    for i in reversed(range(_size)):
      self.L[i] = f(self.L[i + i], self.L[i + i + 1])

  def __delitem__(self, i):
    self[i] = self._def

  def __getitem__(self, i):
    return self.L[i + self._size]

  def __setitem__(self, i, v):
    i += self._size
    self.L[i] = v
    i >>= 1
    while i:
      self.L[i] = self._f(self.L[2 * i], self.L[2 * i + 1])
      i >>= 1

  def __len__(self):
    return self._len

  def query(self, i, acc, x):
    if i >= self._size: #자식이 없는 경우
      return i - self._size
    l = i + i #좌자식
    r = i + i + 1 #우자식
    
    if acc + self.L[l] < x: #찾으려는 수가 좌자식을 합친 것보다 값이 큰 경우
      return self.query(r, acc + self.L[l], x) #우자식으로 탐색
    else :
      return self.query(l, acc, x)


def sol() :
  N = int(input())
  MAX = 2_000_000
  sg = ST([0] * (MAX + 1), 0, lambda x, y: x + y)

  for _ in range(N) :
    q, x = map(int, input().split())
    if q == 1 :
      sg[x] += 1
    elif q == 2 :
      res = sg.query(1, 0, x)
      sg[res] -= 1
      sys.stdout.write(str(res) + "\n")

sol()
```
