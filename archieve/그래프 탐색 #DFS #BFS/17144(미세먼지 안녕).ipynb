```python
import io, os
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def spread(G, D, Q, N, M) :
  while Q :
    x, y = Q.pop()
    cur = G[x][y] #현재 칸의 미세먼지 양
    for dx, dy in [(0, 1), (1, 0), (0, -1), (-1, 0)] :
      nx = x + dx
      ny = y + dy

      if not (0 <= nx < N and 0 <= ny < M): continue
      if G[nx][ny] == -1 : continue

      D[nx][ny] += cur // 5
      D[x][y] -= cur // 5
  
  for i, l in enumerate(D) :
    for j, v in enumerate(l) :
      G[i][j] += v
      D[i][j] = 0

def clean(G, N, M, s, pos) :
  x, y = s
  d = 0
  G[x][y], tmp = 0, G[x][y]
  while True :
    nx = x + pos[d][0]
    ny = y + pos[d][1]

    if not (0 <= nx < N and 0 <= ny < M) :
      d += 1
      continue
    
    if G[nx][ny] == -1 : break
    G[nx][ny], tmp = tmp, G[nx][ny]
    x, y = nx, ny #다음 위치

  
def sol() :
  R, C, T = map(int, input().split())
  G = [[*map(int, input().split())] for _ in range(R)]
  D = [[0] * C for _ in range(R)] #1초 후의 미세먼지 변화량

  q1, q2 = None, None
  for i, l in enumerate(G) :
    for j, v in enumerate(l) :
      if v == -1 :
        if not q1 : q1 = (i, j+1)
        else : q2 = (i, j+1)

  for _ in range(T): 
    Q = []
    for i, l in enumerate(G) :
      for j, v in enumerate(l) :
        if not v or v == -1 : continue
        Q.append((i, j))
    spread(G, D, Q, R, C)
    clean(G, R, C, q1, [(0, 1), (-1, 0), (0, -1), (1, 0)])
    clean(G, R, C, q2, [(0, 1), (1, 0), (0, -1), (-1, 0)])

  print(sum(map(sum, G)) + 2)

sol()
```

- 그냥 문제에서 하라는걸 구현하면 되는 시뮬레이션 문제
- 공기청정기로 미세먼지를 옮기는 과정에서 tmp랑 스왑하는 것과 nx/ny에서 처리되는 부분에서 햇갈려서 조금 시간이 오래걸렸다.
