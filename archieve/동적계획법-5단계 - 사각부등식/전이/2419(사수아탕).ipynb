```python
import io, os, bisect
input=io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

pos = []
def dist(s, e) :
  return abs(pos[s] - pos[e])

MAX = 300
NONE = -1


def sol() :
  N, M = map(int, input().split())
  L = [int(input()) for _ in range(N)]
  L.append(0)
  L.sort()

  def w(l, r, k, w, dp) :
    vis = [[False] * 2 for _ in range(MAX+1) for _ in range(MAX+1)]
    S = [(l, r, k, w)]

    while S: 
      l, r, k, w = S.pop()
      if w == 0 or (l == 0 and r == N) : 
        dp[l][r][k] = 0
        continue
      
      if vis[l][r][k] : continue

      cur = r if k else L
      if l > 0 and dp[l-1][r][0] == NONE :
        S.append((l, r, k, w))
        S.append((l-1, r, 0, w-1))
        continue
      if r < N and dp[l][r+1][1] == NONE :
        S.append((l, r, k, w))
        S.append((l, r+1, 1, w-1))
        continue

      if l > 0 :
        dp[l][r][k] = max(dp[l][r][k], dp[l-1][r][0] - dist(l-1, cur) + M)
      if r < N :
        dp[l][r][k] = max(dp[l][r][k], dp[l][r+1][1] - dist(r+1, cur) + M)
      
      vis[l][r][k] = True
  
  ans = 0
  idx = bisect.bisect_left(L, 0)
  for i in range(N+1) :
    dp = [[[NONE] * 2 for _ in range(MAX+1)] for _ in range(MAX+1)]
    ans = max(ans, w(idx, idx, 0, i, dp))
  print(ans)

sol()
```
