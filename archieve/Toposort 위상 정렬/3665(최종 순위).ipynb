```python
import sys, collections
input = sys.stdin.readline
print = sys.stdout.write

def sol() :
  N = int(input())
  l = [*map(int, input().split())]
  G = [[] for _ in range(N+1)]
  IN = [-1] + [0] * N #진입차수 갯수
  for i, v in enumerate(l):
    G[v] = l[i+1:]
    IN[v] = i
  
  for _ in range(int(input())) :
    a, b = map(int, input().split())
    if a in G[b] :
      G[b].remove(a)
      G[a].append(b)
      IN[a] -= 1
      IN[b] += 1
    else :
      G[a].remove(b)
      G[b].append(a)
      IN[b] -= 1
      IN[a] += 1

  Q = collections.deque()
  for i, e in enumerate(IN[1:], 1) :
    if not e :
      Q.append(i)

  if not Q:
    print('IMPOSSIBLE\n')
    return

  res = []
  while Q:
    v = Q.popleft()
    res.append(v)

    for e in G[v] :
      IN[e] -= 1
      if not IN[e] :
        Q.append(e)

  if len(res) != N :
    print('IMPOSSIBLE\n')
  else :
    print(' '.join(map(str, res))+'\n')

for _ in range(int(input())) :
  sol()
```

- 이상한 템플릿 써보겠다고 고생한 문제.

### 풀이
- 위상정렬을 위한 그래프의 구성을 어떻게 구성하느냐에 대한 문제였다.
- 순위가 뒤바뀌는 것이 아닌, 순위의 `상대적인 위치(위상 순서)`가 역전되는 것이기 때문에 순위의 위상순서를 유지하기 위해선 그래프를 다음과 같이 설정해야 한다.
  - `a - b`를 a가 b보다 순위가 높은 상태라고 할 때, N개의 팀에 대해서
  - 1 - 2, 1 - 3, ... 1 - N
  - 2 - 3, 2 - 4, ... 2 - N
  - ...
  - N-1 - N
  - 위와 같은 형태로 구성해야 한다.
- 상대적인 등수가 바뀌는 것은 연결 정보를 뒤집는 것으로 구현한다.
- 이후 위상정렬을 진행하면 된다.
