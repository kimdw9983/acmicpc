```python
import sys
input=sys.stdin.readline

N = int(input())

l = []
for _ in range(N) :
  l.append(int(input()))
S = [[l[0], 1]]

answer = 0
#0이 들어가는가???
for i, v in enumerate(l[1:]) :
  # print(answer, S)
  if S[-1][0] < v : #top보다 클경우
    x = 0
    while S and S[-1][0] < v :
      x += S.pop()[1]
    answer += x

    #top 밑에 뭐가 더 있는지 확인
    if S :
      answer += 1 if S[-1][0] > v else S[-1][1]
    
    if S and S[-1][0] == v :
      S[-1][1] += 1

      tmp = S.pop() #top 밑에 뭐가 더 있는지 확인
      answer += 1 if S else 0
      S.append(tmp)
    else :
      S.append([v, 1])
  elif S[-1][0] == v :
    answer += S[-1][1]
    
    #top 밑에 뭐가 더 있는지 확인
    tmp = S.pop()
    answer += 1 if S else 0
    S.append(tmp)

    S[-1][1] += 1
  elif S[-1][0] > v : #top보다 작은 경우
    answer += 1
    S.append([v, 1])
  else :
    assert(False)

# print(answer, S)
print(answer)
```
