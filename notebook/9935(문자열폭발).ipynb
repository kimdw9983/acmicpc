```python
import sys
input = sys.stdin.readline
sys.setrecursionlimit(10**6)

S = input().rstrip()
exp = input().rstrip()
first = exp[0]

stack = []

def r(start) :
  if start >= len(S) : 
    return (len(S), False)
  
  i = start + 1
  stack.append(first)

  exp_i = 1
  while i < len(S) : #여기서의 i는 확인이 안된 i이다.
    #첫 번째 문자를 발견
    if S[i] == first :
      # stack.append(first) 이 작업은 r()에 들어가자마자 실행된다.
      result = r(i)
      if result[1] is False :
        #실패했을 경우
        return (result[0], False) #확인이 안된 문자열의 i를 반환
      else :
        #성공했을 경우
        i = result[0] #확인이 안된 i
    elif S[i] == exp[exp_i] :
      #일치하는 문자를 발견
      stack.append(S[i])
      if S[i] == exp[-1] : #그것이 마지막 문자일 경우(현재 스택의 모든 문자가 일치한 상태)
        for _ in exp :
          stack.pop()
        return (i + 1, True) #현재 문자(i)는 확인한 상태이므로 i+1을 반환, 한 스택 나온다
      else : #마지막 문자가 아닌경우
        i += 1
        exp_i += 1
    else :
      #일치하지 않은 문자를 발견
      stack.append(S[i]) #출력을 해야 하니까 넣음. 
      return (i + 1, False) #i는 확인이 됐으므로 확인이 안된 i + 1을 반환
  return (len(S), False) #index 초과
  

p = 0
answer = []
if len(exp) == 1 :
  answer.extend(S.replace(exp, ''))
else :
  while p < len(S) :
    if S[p] == first :
      result = r(p)
      if result[1] is False :
        answer.extend(stack)
        stack.clear()
      p = result[0]
    else :
      answer.extend(S[p])
      p += 1

if answer :
  sys.stdout.write("".join(answer) + "".join(stack) + "\n")
else :
  print("FRULA")
```
