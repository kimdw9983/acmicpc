## Greedy Stays ahead
- 탐욕 알고리즘이 항상 좋다(앞선다)는 방식으로 분석하는 방법. 이 문제의 대표주자 격으로 Interval Scheduling이 있다. 증명하는 흐름은 다음과 같다.
  - 그리디 알고리즘 `G`를 하나 제시한다. 그리고 또 다른 최적해를 도출하는 가상의 알고리즘 `X*`도 제시한다.
  - 만약 `G`의 선택이 `X*`의 선택보다 항상(어떤 기준에 따라) 좋다면, 전체적으로 봤을 때에도 탐욕 알고리즘의 선택이 최적해를 도출한다는 것을 증명할 수 있다.

### Interval Scheduling Problem(회의실 배정 문제)
- 최대한 많은 회의를 배정하는 문제. 이에 대한 해법은 잘 알려진 방법은 다음과 같다.
- > 모든 회의를 끝나는 시각의 오름차순으로 정렬한 후, 각 회의를 순서대로 고려하면서 회의실을 배정할 수 있으면 넣고 그렇지 않으면 버린다.
- 먼저 끝나는 회의부터 배정한다면 회의가 끝난 후 다른 회의를 처리할 수 있는 시간이 더 길어지기 때문에 좋을 것이다. 이를 `G`라고 해보자.
- 어떤 작업 $p$ 에 대해, 이 작업이 끝나는 시간을 $s(p)$ , 끝나는 시간을 $f(p)$ 라고 하자. \
그리고 $G$가 배정한 작업의 집합을 $S := \{p_1, p_2, ..., p_k\}$ 라고 하자. 이때 $p_1, p_2, ..., p_k$ 는 회의실 배정에 처리되는 순서이다. 즉
- $s(p_1) \leq f(p_1) \leq s(p_2) \leq f(p_2) \leq ... \leq s(p_k) \leq f(p_k)$ 를 만족한다.
  - 같은 회의에선 시작시간보다 끝시간이 더 크며, 다음 회의를 배정하기 전, 즉 다음 회의의 시작시간은 이전 회의의 끝시간보다 크거나 같다.
- 비교를 위해 최적해 $O = \{q_1, q_2, ... , q_l\}$ 을 하나 가져와보자. 여기서도 마찬가지로 $q_1, q_2, ..., q_l$ 는 회의실 배정에 처리되는 순서로써
- $s(q_1) \leq f(q_1) \leq s(q_2) \leq f(q_2) \leq ... \leq s(q_l) \leq f(q_l)$ 를 만족한다.
- 여기서 보여야 할 것은 $k \geq l$ 이다. 즉, $G$가 배정한 작업의 수가 최적해의 작업의 수보다 크거나 같다는 것이다.

#### Lemma1. Correctness by greedy-stays-ahead
> 모든 $i = 1, ..., l$ 에 대해, $p_i \in S$ 이며, $f(p_i) \leq f(q_i)$ 를 만족한다.
- 최적해의 모든 원소가 $S$ 에도 포함되어있고, $S$ 에 포함된 원소들은 최적해의 원소들보다 빨리 끝난다.
- 최적해의 각 $q_i \in O$ 를 $p_i \in S$ 에 짝지어줄 수 있고, 그러면서 항상 $p_i$ 가 $q_i$ 보다는 늦지 않게 끝난다.
- 즉, 가상의 최적 알고리즘이 있다고 했을 때, 이것이 선택하는 작업보다 탐욕 알고리즘이 그때 선택하는 작업이 더 먼저 끝난다.
- 그리고 $O$ 의 모든 작업을 $S$ 에 하나씩 짝지어주었으므로, $S$ 가 최소 $O$ 의 개수 만큼은 회의실을 배정했다.

#### lemma 1 증명(수학적 귀납법). 
- $i = 1$ 일 때 `G`는 주어진 작업들 중 가장 일찍 끝나는 회의를 배정할 것이므로 분명 $p_1$ 이 존재하고 $f(p_1) \leq f(q_1)$ 를 만족한다.
- 다음으로 $j > 1$ 에 대해 $i = j - 1$ 에 대해 정리가 성립한다고 가정했을 때 $p_j$ 가 $S$ 에 존재하고 $f(p_j) \leq f(q_j)$ 를 만족한다고 하자.
  - 만약 그러한 $p_j$ 가 존재하지 않는다면 $p_{j-1}$ 을 고른 이후 더이상 할당할 작업이 없다는 의미가 된다.
  - 하지만 $O$ 에 포함된 $q_i$ 를 생각해보면 이것은 모순이다. $O$ 의 작업들도 문제의 정의에 의해 서로 작업시간이 겹치지 않아야 하므로 $f(q_{j-1}) \leq s(q_j)$ 를 만족한다.
- 여기서 가정에 의해 $f(p_{j-1}) \leq f(q_{j-1})$ 이 되어 $f(p_{j-1}) \leq s(q_j)$ 를 만족한다. ( $f(p_{j-1}) \leq f(q_{j-1}) \leq s(q_j)$ )
- 이는 `G`가 $p_{j-1}$ 을 고른 이후에도 $q_j$ 를 고를 수 있다는 의미이다. 따라서 모순이다.
- 
- 다음은 $f(p_j) \leq f(q_j)$ 를 만족하는지를 보이자.
- 위에서 보였던 것처럼 탐욕 알고리즘이 $p_{j-1}$ 을 선택한 후 $p_j$ 를 고를 수 있었다. 그런데 만약 $f(p_j) > f(q_j)$ 라면 `G`는 $p_j$ 가 아닌 $q_j$ 를 골랐어야 한다. 따라서 모순이 된다.

## Certificate argument(증거 논의)
- 알고리즘이 답과 함께 그 답이 최적해라는 증거를 함께 출력하도록 하는 방법.

### 강의실 배정 문제
- Interval Scheduling 문제의 변형으로, 이번엔 모든 회의를 최소한의 강의실을 사용해서 배정하고자 하는 문제이다.
- `작업의 집합` $P$ 가 주워졌을 때 각 `직업` $p \in P$ 는 시작시간 $s_p$ 와 종료시간 $f_p$ 를 가지고 있다.
- 한 강의실에서 동시에 두 개 이상의 직업이 진행될 수 없다. 단, 한 작업이 끝날 때 바로 다음 작업을 수행시킬 순 있다.
  - 이러한 작업의 수행 시간을 $[s(p), f(p))$ 라고 표현할 수 있다.
- 다음을 만족하는 $\mathcal{S} = \{S_1, ..., S_k\}$ 중 크기 $k$ 가 가장 작은 것을 찾아야 한다.
  - 각 $S_j \in \mathcal{S}$ 는 서로 겹치지 않은 $P$ 의 부분집합이다.\
  즉, 임의의 서로 다른 두 작업 $p, p' \in S_j$ 에 대해, $[s(p), f(p)) \cap [s(p'), f(p')) = \emptyset$ 이다.
- 이때 최소한 몇개의 강의실이 필요할까? 해답의 `lower_bound`를 생각해보자면, 모든 작업을 수행해야 하므로 `어떤 시간에 동시에 수행되는 작업의 개수 보다`는 많아야 한다.\
다시 말해, 어떤 시각 $t$ 에 수행되는 작업의 집합을 $P_t := \{p \in P : t \in [s(p), f(p)) \}$ 라고 할때 다음 lemma 2를 만족해야 한다.

#### Lemma 2. Certificate(증거)
> $P$ 가 어떻게 주어지든 간에 모든 회의를 배정하기 위해서는 임의의 시각 $t$ 에 대해 $|P_t|$ 이상의 강의실이 필요하다. \
다시말해, $\cal S$ 가 올바른 답(feasible solution) 이라면, 반드시 $|\mathcal{S}| \geq \max_t|P_t|$ 를 만족한다.

- $|\mathcal{S}| \geq \max_t|P_t|$ : 모든 시간 $t$ 에 대해서 강의가 가장 많이 겹치는 시간은 $\cal|S|$ 보다 작거나 같다. 즉 동시에 일어날 수 있는 가장 많은 회의의 수를 의미한다.
- 이때 다음과 같은 lemma 3을 이끌어낼 수 있다.

#### Lemma 3. Optimality through certtificate
> 만약 알고리즘이 올바른 답 $\mathcal S$ 와 함께 $|\mathcal{S}| = |P_{t*}|$ 인 어떤 시각 $t*$ 를 찾는다면, $\cal S$ 는 최적해이다.

- 문제의 최적해를 $\cal O$ 라고 할 때, 이 답도 feasible solution 이므로, Lemma 2에 의해 다음을 만족한다
  - $|\mathcal O| \geq \max_t|P_t| \geq |P_{t*}| = |\mathcal S|$ 
- $\cal O$ 최적해라고 했는데, $\cal S$ 가 그것보다 크기가 더 작거나 같으므로 $\cal S$ 는 최적해이다.
- 다시 말해, 알고리즘이 $\cal S$ 와 함께 Lemma 3을 만족하는 $t*$ 를 `증거`로 제출하면, 알고리즘이 최적의 답을 출력한다는 것을 보일 수 있다.
- >시간의 흐름에 따라 다음을 고려한다. 매 시각 $t$ 에 시작하는 작업이 있을 때 지금까지 배정한 회의실에 배정할 수 있으면 그것들 중 아무 것에 할당한다.\
현재 배정한 회의실만으로 불가능하다면, 새로운 회의실을 만들고 거기에 할당한 다음 $t* \leftarrow t$ 를 기억한다.
  - 
  - 매 시각마다 $t*$ 는 $|\cal S| = |P_{t*}$ 를 만족한다는 것을 확인할 수 있고, 이로써 알고리즘이 원하는 대로 동작함을 증명할 수 있다.

## Exchange Argument (교환 논의)
- 가상의 최적해를 하나 고정하고, 이 최적해를 약간씩 수정하여 알고리즘의 답과 동일하게 하는 방법
  - 여기서 매번 최적해를 수정할 때 답이 나빠지지 않는다는 것을 보인다면, \
  최종적으로 알고리즘의 답이 최적해보다 나쁘지 않다는 것을 보이게 되므로, 그 답 역시 최적해와 같다는 것을 보일 수 있다.

### 소가 길을 건너간 이유 4(기계와 작업으로 대체)
- 처리해야할 작업들이 있고, 각 작업은 생성 시각과 마감 시각이 주어진다. 
  - 생성하기 전에는 작업을 처리할 수 없으며, 생성된 후에는 마감 시각 전까지 기계에 할당하지 못하면 작업은 처리되지 못한다.
- 이번엔 공용 기계의 도움을 받을 수 있다.
  - 특정한 시각에 최대 하나의 작업을 처리할 수 있다고 신호를 보내준다.
  - 즉, 그 시각에 처리 가능한 작업 하나를 공용 기계에 보내는 것이다. 한번 작업을 넘기게 되면 그 일은 끝난 것으로 생각하면 된다.
- 최대한 많은 작업을 공용 기계로 보내는 것이 목표이다.

- 이 문제를 해결할 가상의 최적해를 반환하는 알고리즘을 구상해보자. 어떤 작업을 보내야 할까? 최대한 많은 작업을 공용 기계로 보내야 한다.
  - 곧 마감되는 작업을 보내는 것이 이득일 것이다.
  - 마감 시간이 늦는 작업을 보내게 된다면, 그것보다 일찍 마감되는일이 다 끝나고 비는 시간이 생길 수 있기 때문이다.
  - 따라서 이 기준으로 최적해를 뽑는 가상의 알고리즘 `G`는 다음과 같이 정의할 수 있다.
> 매번 공용 기계로부터 신호를 받으면, 처리할 수 있는 작업 중 마감시각이 이른 것을 공용 기계로 전송한다.\
만약 처리할 수 있는 작업이 없다면 아무 행동도 하지 않는다.

#### 문제의 정의
- 작업의 집합 $P$ 가 주어지고, 각 작업 $p \in P$ 에 대해서 생성 시각 $r(p)$ 와 마감 시각 $d(p)$ 가 주어진다.
- 공용 기계가 주는 시각의 집합을 $T$ 라고 하고, 그 신호의 개수를 $n := |T|$ 라고 한다.
- 어떤 작업이든 그것을 처리하려면 신호가 들어온 시각에 활성화 돼있어야 한다.
- 모든 작업은 최대 한 번만 수행하면 충분하고, 한 신호에는 최대 하나의 작업을 처리할 수 있다.
- 이 문제의 올바른 해는 다음을 만족하는 작업-신호 쌍의 집합 $M \subseteq P \times T$ 가 된다.
  - 임의의 작업-신호 쌍 $(p, t) \in M$ 은 $r(p) \leq t \leq d(p)$ 를 만족한다.
  - 임의의 작업 $p \in P$ 나 신호 시각 $t \in T$ 나 $M$ 에는 최대 한번만 등장한다.\
  예를 들어 $t_1 \ne t_2$ 에 대해서 $(p, t_1) \in M$ 이면서 동시에 $(p, t_2) \in M$ 일 수 없으며, $p_1 \ne p_2$ 에 대해 동시에 $(p_1, t) \in M$ 이면서 동시에 $(p_2, t) \in M$ 일 수 없다.
- 그러한 $M$ 중에서 크기가 가장 큰 것을 찾는 것이다
