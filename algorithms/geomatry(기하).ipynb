### 극좌표계 -> 직교좌표계 변환
- 좌표 회전 method
- 25308(방사형 그래프)

```python
x1 = 5
y1 = 9
degree = 30

radian = math.radians(degree)
x2 = x1 * math.cos(radian)
y2 = y1 * math.sin(radian)
print(f'x1, y1({x1}, {y1})을 {degree}도 회전시킨 좌표는 ({x2}, {y2})')
```
