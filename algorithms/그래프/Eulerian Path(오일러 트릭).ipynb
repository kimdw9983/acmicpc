### 오일러 경로와 오일러 회로(Eulerian Path and Eulerian Circuit)
- 유향 혹은 무향 그래프가 주워졌을 때, 그래프에 존재하는 `모든 간선을 정확히 1번만 방문`하는 연속된 경로.
- 한붓 그리기를 떠올리면 된다.
- 오일러 회로는 오일러 경로의 일종으로, `시작점과 끝점이 같은 경로`를 말한다.
- 
  - 위 그림에서 오일러 회로가 존재하고, 그 경로중 하나는 `[A, B, C, E, F, C, D, A]`이다.

### 오일러 경로와 오일러 회로의 성립 조건
- 오일러 경로의 존재 여부를 판단하는 방법은 다음과 같다.
  - 무향 그래프의 경우 차수(degree)가 홀수인 정점이 0개(이 경우 오일러 회로) 또는 2개 있어야 한다.
    - 홀수인 차수가 0개인 경우, 어떤 점을 시작점으로 하더라도 오일러 회로가 존재한다.
    - 2개인 경우, 그 2개의 정점을 각각 시작점과 끝점으로 해야 오일러 경로가 만들어진다.
    - 시작점과 끝점을 제외하고서는 `들어오는 간선`이 존재할 때 `나가는 간선`이 존재해야 하기 때문에 항상 차수가 2의 배수꼴로 붙어야 한다. (필요충분조건?)
  - 유향 그래프에서 오일러 회로가 존재하려면 모든 정점에 대해 진입차수와 진출차수가 일치(이 경우 오일러 회로)하거나, \
  시작점과 끝점을 제외하고 모두 진입차수와 진출차수가 일치해야 하며, 진입차수가 1개 적은 정점이 시작점이면서 진출차수가 1개 적은 정점이 끝점이여야 한다.

### Hierholzer's Algorithm
- 오일러 회로를 찾는 효율적인 알고리즘
1. 임의의 정점 $V$ 를 잡고 dfs를 돌려서 $V$ 로 되돌아오는 사이클을 찾는다. 그러한 경로를 $P$ 라고한다.
2. 그래프의 모든 정점을 탐색하지 않았다면, $P$ 에서 탐색하지 않은 정점이 있는 정점 $V'$ 를 찾아 그 정점부터 다시 dfs를 돌린다. 그 경로를 $P'$ 라고 한다.
    - (구현) 이때 방문하지 않은 간선이 존재하는 $V'$를 찾을 때 까지 거꾸로 출력할 수 있다.
3. $P$ 와 $P'$ 를 합친다.
4. 모든 정점을 탐색할 때 까지 2, 3을 반복한다.
