### 대수적 체계 (Algebraic System)
- 원소의 `집합`과 `원소`들 간의 연산으로 구성된 체계
- 대수적 체계의 조건
  - 원소의 집합이 있어야 한다.
  - 원소들 간의 연산이 있어야 한다.
  - 연산에 대한 닫힘성이 있어야 한다.
  - 연산에 대한 결합법칙이 성립해야 한다.
  - 연산에 대한 항등원이 있어야 한다.
  - 연산에 대한 역원이 있어야 한다.
- 대수적(Algebraic)이라 함은, 다항식의 근과 관련된 내용을 말한다.
  - 관련 내용으로 `5차 이상의 다항식은 근의 공식이 존재하지 않는다` 등이 있다. (Abel-Ruffini Theorem)

### 군 (Group)
- 한개의 연산으로 구성된 대수적 체계

### Galois Theory(갈루아 이론)
- 체의 대칭성 정보를 `손실 없이` 군으로 가져와서, 그 군의 구조를 이용하여 체의 구조를 연구하는 분야
- Galois Field: 유한개의 원소를 가지고, 그 안에서 대수적 구조를 형성하는 체
  - 유한체(Finite Field) 라고도 불린다.
- 모듈러계와 비슷하다. 거기에 몇몇 약속된 연산에 대한 규칙도 명시하는 방법이다.
