- 통일성있고 간결한 문서를 만들기위한 나만의 규칙.
- 의식의 흐름을 적었더라도, 그것은 중요한 내용일 수 있고, 나중에도 이해할 수 있어야 하기 때문.

### 중요한 것

- 한 문서 내에서, 윗 부분에서 정의한 내용은 아랫 부분에서도 동일하게 적용된다.\
그리고 문제에서 주어진 변수명은, 문서에 가장 위에 정의한 것으로 간주한다.
  - 문제에서 쓰이는 표현을 재정의하지 말 것.
  - 불필요한 정의의 반복을 줄일 것.
  - 다른 논리를 설명하고 싶다면 다른 표현을 사용해 충돌이 일어나지 않도록 설정할 것.
- 특수한 문제가 아닐경우 설명은 1-index, 구현속의 comment은 0-index를 기준으로 한다.
  - 어짜피 둘다 익숙해져야 하니 통일하지 않는다. 
  - 일종의, off-by-one 연습.
- 구현과 관련된 내용은 `로 감싸고, 수학 혹은 formal하게 표현하는 것과 관련된 내용은 $로 감싼다.
  - ex) `A[:4]` 는 A의 0~3번째, $A[:4]$ 는 A의 1~4번째
- 문제를 풀고 나서 정리하자.
  - 자꾸 생각나는 것을 먼저 정리하려고 시간을 쓰면, 문제 푸는 시간이 늘어난다. 문제를 풀 땐 문제만 풀자.
  - 필요한 내용이 있다면, 메모를 해둘 것.

### 조금 덜 지켜도 되는 것
- 적어도, 한 문서내에서 쓰이는 표현에 대한 띄어쓰기는 통일한다.
  - 예를들어 나의 국평오 이슈로인해 띄어쓰기를 잘 모르는 합성어를 표현해야 한다고 하자. 
  - 대충 `부분문자열` 이라는 단어라고 치자.
  - 올바른 띄어쓰기가 뭐던간, 한 문서내에서 `부분 문자열` 이랑 `부분문자열` 이랑 같이 쓰면 안된다는 이야기. 
- 풀이 내에 코드를 여러번 적을 때 가급적 반복되는 부분은 지운다.
